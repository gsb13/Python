## üîç What is Logging in Python?

**Logging** is a way to track events that happen when your software runs. Python's `logging` module lets you report status, errors, and debug messages to different outputs (like console, files, emails, etc.).

It‚Äôs **better than using print()** because:

* It can log at different **severity levels**.
* You can **write to files** or other destinations.
* You can **turn logging off** or **change its level** without changing your code.

---

## üß± Logging Levels

| Level      | Function                             | Numeric Value |
| ---------- | ------------------------------------ | ------------- |
| `DEBUG`    | Detailed info (for developers)       | 10            |
| `INFO`     | Confirmation that things are working | 20            |
| `WARNING`  | Something unexpected, not serious    | 30            |
| `ERROR`    | A serious problem                    | 40            |
| `CRITICAL` | Very serious error                   | 50            |

## ‚úÖ Basic Logging Example

### üî∏Program 1: Basic Console Logging

```python

# Set the logging level to DEBUG

```


## ‚öôÔ∏è Multiple Handlers (Console + File)

### üî∏Program 5: Logging to Both Console and File

```python
import logging

# Create a logger object
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.DEBUG)

# Create file handler
file_handler = logging.FileHandler("multi.log")
file_handler.setLevel(logging.DEBUG)

# Create console handler
console_handler = logging.StreamHandler()
console_handler.setLevel(logging.WARNING)

# Create formatter
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')

# Add formatter to both handlers
file_handler.setFormatter(formatter)
console_handler.setFormatter(formatter)

# Add handlers to the logger
logger.addHandler(file_handler)
logger.addHandler(console_handler)

logger.debug("Debug message")    # Goes to file only
logger.warning("Warning message")  # Goes to both
```

### üîπConsole Output:

```
2025-09-18 12:02:00,123 - WARNING - Warning message
```

### üîπFile `multi.log`:

```
2025-09-18 12:02:00,122 - DEBUG - Debug message
2025-09-18 12:02:00,123 - WARNING - Warning message
```

## üî∞ Python Logging Interview Questions (with Answers)

### **1. What is the Python `logging` module, and why is it preferred over `print()` statements?**

**Answer:**
The `logging` module provides a flexible framework for emitting log messages from Python programs. Unlike `print()`, which only outputs to the console, logging can:

* Write to files, sockets, or streams.
* Use severity levels (DEBUG, INFO, WARNING, ERROR, CRITICAL).
* Be configured to change logging levels and formats dynamically.
* Be turned off or redirected without code changes.

This makes logging more suitable for production and debugging in real-world applications.

---

### **2. What are the different logging levels in Python, and when should you use each?**

**Answer:**

| Level      | Use Case                                                                  |
| ---------- | ------------------------------------------------------------------------- |
| `DEBUG`    | Detailed information, used during development.                            |
| `INFO`     | Confirmation that things are working as expected.                         |
| `WARNING`  | An indication something unexpected happened, but program still works.     |
| `ERROR`    | A more serious issue that prevented part of the program from functioning. |
| `CRITICAL` | A serious error, indicating the program may not continue running.         |

---

### **3. How do you log messages to both a file and the console at the same time?**

**Answer:**
You use multiple **handlers** (e.g., `FileHandler`, `StreamHandler`) and add them to the logger:

```python
import logging

logger = logging.getLogger("AppLogger")
logger.setLevel(logging.DEBUG)

file_handler = logging.FileHandler("app.log")
console_handler = logging.StreamHandler()

formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
file_handler.setFormatter(formatter)
console_handler.setFormatter(formatter)

logger.addHandler(file_handler)
logger.addHandler(console_handler)

logger.info("This will go to file and console")
```

---

### **4. What is the use of `logging.getLogger(name)`?**

**Answer:**
`logging.getLogger(name)` retrieves a logger with the specified name:

* Named loggers are useful for **module-level logging** in large applications.
* You can configure different loggers differently.
* It follows a **hierarchical naming** (e.g., `app.module.submodule`).

Calling `getLogger()` without a name gives you the **root logger**.

---

### **5. Can you change the logging level at runtime? How?**

**Answer:**
Yes, you can change the logging level using `setLevel()`:

```python
logger = logging.getLogger()
logger.setLevel(logging.WARNING)  # Now only WARNING and above are logged
```

You can even set different levels for different handlers.

---

### **6. What is the purpose of `basicConfig()`? What are its limitations?**

**Answer:**
`logging.basicConfig()` is a quick way to configure logging:

* Sets default level, format, and optionally a filename.
* Must be called **only once** (usually at the beginning).

**Limitations:**

* It does not allow multiple handlers directly.
* If `getLogger()` is used before `basicConfig()`, it may not work as expected.

For complex logging, it's better to configure handlers manually or use `dictConfig()`.

---

### **7. How do you handle exceptions using logging?**

**Answer:**
Use `logger.exception()` inside an exception block. It logs the error **with the traceback**:

```python
import logging

logging.basicConfig(level=logging.ERROR)

try:
    1 / 0
except ZeroDivisionError:
    logging.exception("Division by zero error occurred")
```

This automatically includes the stack trace in the log.

---

### **8. What is a `Handler` and a `Formatter` in logging?**

**Answer:**

* **Handler**: Sends log messages to a destination (e.g., console, file, email). Examples: `StreamHandler`, `FileHandler`, `SMTPHandler`.
* **Formatter**: Defines the layout of the log message (timestamp, level, message).

Handlers and formatters allow flexible logging configurations.

---

### **9. How would you implement rotating log files (e.g., daily logs or size-based rotation)?**

**Answer:**
Use `RotatingFileHandler` or `TimedRotatingFileHandler`.

**Example (size-based rotation):**

```python
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("app.log", maxBytes=2000, backupCount=5)
```

**Example (daily rotation):**

```python
from logging.handlers import TimedRotatingFileHandler

handler = TimedRotatingFileHandler("daily.log", when="midnight", backupCount=7)
```

---

### **10. What is the difference between `Logger.debug()` and `logging.debug()`?**

**Answer:**

* `logging.debug()` uses the **root logger**.
* `Logger.debug()` is called on a **specific logger object** (created using `getLogger()`).

Using named loggers is better in large applications for modular control.

## üß† Bonus Conceptual Questions

### **11. Can logging slow down your application?**

**Answer:** Yes, if misused. Excessive or synchronous logging (especially in high-frequency loops or I/O operations) can become a bottleneck. Solutions include:

* Logging only necessary information.
* Using asynchronous logging (e.g., `QueueHandler`).
* Limiting log levels in production.

---

### **12. How do you load logging configuration from a file?**

**Answer:** Use `logging.config` module.

**Example (YAML or dictConfig):**

```python
import logging.config

logging.config.fileConfig('logging.conf')
```

Or:

```python
logging.config.dictConfig(config_dict)
```
