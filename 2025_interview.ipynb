In Python, `super()` is a built-in function used primarily in **class inheritance** to refer to the **parent (or superclass)** of the current class. It allows you to call methods or access properties from the superclass **without explicitly naming it**, which is especially useful in complex hierarchies or with multiple inheritance.

---

### ✅ Why Use `super()`?

* To avoid hardcoding the parent class name.
* To support **cooperative multiple inheritance**.
* To make code **easier to maintain and extend**.

---

### 🔹 Basic Syntax

```python
super().method_name(arguments)
```

---

### 🔸 Example 1: Single Inheritance

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        super().speak()  # Call the parent class method
        print("Dog barks")

d = Dog()
d.speak()
```

**Output:**

```
Animal speaks
Dog barks
```

---

### 🔸 Example 2: `__init__` with `super()`

```python
class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, name, employee_id):
        super().__init__(name)  # Call the parent constructor
        self.employee_id = employee_id

e = Employee("Alice", 123)
print(e.name, e.employee_id)
```

**Output:**

```
Alice 123
```

---

### 🔸 Example 3: Multiple Inheritance

```python
class A:
    def do_something(self):
        print("A")

class B(A):
    def do_something(self):
        super().do_something()
        print("B")

class C(A):
    def do_something(self):
        super().do_something()
        print("C")

class D(B, C):
    def do_something(self):
        super().do_something()
        print("D")

d = D()
d.do_something()
```

**Output (due to MRO - Method Resolution Order):**

```
A
C
B
D
```

---

### 🧠 Notes

* `super()` only works in new-style classes (i.e., classes that inherit from `object`), which is default in Python 3.
* You don’t need to pass `self` or class name explicitly like `super(ClassName, self)` in Python 3 — just `super()` works.

````python
def is_balanced(s):
    stack = []
    bracket_map = {')': '(', ']': '[', '}': '{'}
    
    for char in s:
        if char in bracket_map.values():  # Opening brackets
            stack.append(char)
        elif char in bracket_map:         # Closing brackets
            if not stack or stack[-1] != bracket_map[char]:
                return False
            stack.pop()
    
    return not stack 
print(is_balanced("()[]{}"))  # Output: True
print(is_balanced("()[]{")) 
````

````python
from collections.abc import Iterable

def flatten_gen(lst):
    for item in lst:
        if isinstance(item, Iterable) and not isinstance(item, (str, bytes)):
            yield from flatten_gen(item)
        else:
            yield item

# Usage:
print(list(flatten_gen([[1, [2, 3]], 4])))
````

````python
from collections.abc import Iterable

def flatten(obj):
    result = []

    def _flatten(item):
        if item is None:
            return
        if isinstance(item, Iterable) and not isinstance(item, (str, bytes)):
            for sub in item:
                _flatten(sub)
        else:
            result.append(item)

    _flatten(obj)
    return result
````

````python
print(flatten([[1, [2, 3]], 4]))                     # → [1, 2, 3, 4]
print(flatten([1, (2, 3), {4, 5}, "abc"]))           # → [1, 2, 3, 4, 5, 'abc']
print(flatten(["abc", [1, [2, None], (3, 4)]]))      # → ['abc', 1, 2, None, 3, 4]
print(flatten([None, [[], [5, 6]]]))   
````

````python
# write a function to check if a string containing brackets is balanced — 
# meaning every opening bracket has a corresponding closing one in the correct order.

# example: "()", "[]", "{[()]}", "({[]})" are balanced

def is_balanced(string):
    stack = []
    balance = {'(': ')', '[': ']', '{': '}'}
    
    for char in string:
        if char in balance:  # If it's an opening bracket
            stack.append(char)
        elif char in balance.values():  # If it's a closing bracket
            if not stack or balance[stack.pop()] != char:
                return False
    
    return not stack  # If stack is empty, it's balanced

# Example usage:
input_str = '()[]{}'
print(is_balanced(input_str))  # Output: True
````

A **decorator** in programming—particularly in **Python**—is a design pattern that allows you to **add behavior to a function or class without modifying its actual code**. It’s a powerful feature used for **code reuse, separation of concerns, and enhancement of functionality** in a clean, readable way.

---

### 🔧 What Is a Decorator?

In Python, a decorator is a **function that takes another function as an argument**, adds some functionality to it, and returns it.

#### Example:

```python
def decorator_function(original_function):
    def wrapper_function():
        print("Wrapper executed before", original_function.__name__)
        return original_function()
    return wrapper_function

@decorator_function
def say_hello():
    print("Hello!")

say_hello()
```

**Output:**

```
Wrapper executed before say_hello
Hello!
```

---

### 🎯 Significance of Decorators

1. **Separation of concerns**: Keep core logic separate from auxiliary tasks like logging or access control.
2. **Code reusability**: Reuse the same functionality across many functions or methods.
3. **Cleaner syntax**: The `@decorator_name` syntax makes code more readable and maintainable.
4. **DRY principle**: Helps follow "Don't Repeat Yourself" by wrapping repetitive logic.

---

### 🌐 Real-Time Use Cases

#### 1. **Authentication in Web Applications**

Used in frameworks like Flask or Django.

```python
from flask import Flask, request
app = Flask(__name__)

def login_required(f):
    def wrapper(*args, **kwargs):
        if not request.args.get("user"):
            return "Access Denied"
        return f(*args, **kwargs)
    return wrapper

@app.route('/dashboard')
@login_required
def dashboard():
    return "Welcome to Dashboard"
```

#### 2. **Logging**

Used to track function execution without modifying the function itself.

```python
def log_function_call(func):
    def wrapper(*args, **kwargs):
        print(f"Calling function {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@log_function_call
def process_data():
    print("Processing data...")
```

#### 3. **Caching Results**

Decorators can cache results of expensive functions.

```python
from functools import lru_cache

```

#### 4. **Timing Function Execution**

Used to benchmark or profile code.

````python
# Decorator to cache results
def cache_decorator(func):  #function to cache results
    """Decorator to cache results of a function."""
    cache_dict = {}

    def wrapper(num):#wrapper function to handle caching having argument num
        if num in cache_dict:
            print(f"Returning cached result for {num}")
            print(cache_dict)
            return cache_dict[num]
        else:
            result = func(num)
            cache_dict[num] = result
            print(f"Caching result for {num}")
            print(cache_dict)
            return result
        
    return wrapper

# Use the decorator
@cache_decorator
def square(num):
    return num * num

# Test calls
print(square(10))  # Should calculate and cache
print(square(10))  # Should return cached result
print(square(5))   # Should calculate and cache
print(square(5))   # Should return cached result
print(square(20))  # Should calculate and cache
````

A **generator** in Python is a special type of **iterator** that allows you to iterate over data **without storing the entire dataset in memory**. It’s ideal for working with **large datasets, streaming data, or infinite sequences**.

---

### 🔁 What Is a Generator?

A generator is defined like a normal function but uses the **`yield`** keyword to return values one at a time, **pausing between each**.

#### Example:

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(5):
    print(num)
```

**Output:**

```
1
2
3
4
5
```

The key here is that **`yield` suspends the function’s state**, so it can resume where it left off.

---

### 🎯 Significance of Generators

| Feature              | Benefit                                     |
| -------------------- | ------------------------------------------- |
| **Memory efficient** | Doesn't load all data at once into memory   |
| **Lazy evaluation**  | Computes values on the fly                  |
| **Performance**      | Useful for large or infinite sequences      |
| **Clean syntax**     | Easier than writing custom iterator classes |

---

### 🌐 Real-Time Use Cases of Generators

#### 1. **Reading Large Files Line by Line**

Instead of loading the entire file into memory:

```python
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_large_file("huge_log.txt"):
    print(line)
```

#### 2. **Streaming Data (APIs, IoT, Logs)**

```python
import time

def stream_sensor_data():
    while True:
        yield {"temperature": 25 + (time.time() % 5)}
        time.sleep(1)

for data in stream_sensor_data():
    print(data)
```

#### 3. **Producing Infinite Sequences**

```python
def infinite_counter():
    num = 0
    while True:
        yield num
        num += 1

for i in infinite_counter():
    if i > 10:
        break
    print(i)
```

#### 4. **Database Pagination (Lazy Query Results)**

You can yield rows from a database query without loading all at once.

```python
def fetch_in_batches(cursor, batch_size=100):
    while True:
        records = cursor.fetchmany(batch_size)
        if not records:
            break
        yield from records
```

#### 5. **Pipeline Processing**

Split a large process into stages that pass data using generators.

```python
def tokenize(texts):
    for text in texts:
        yield text.split()

def lowercase(words_stream):
    for words in words_stream:
        yield [w.lower() for w in words]

tokens = lowercase(tokenize(text_list))
for item in tokens:
    print(item)
```

### ✅ Summary

| Concept      | Description                                  |
| ------------ | -------------------------------------------- |
| `yield`      | Suspends function execution, returns value   |
| Memory-safe  | Useful for large datasets                    |
| Lazy loading | Only computes what’s needed when it’s needed |

````python
# write fibonacci series without using generator
def fibonacci(n):
    a, b = 0, 1
    result = []
    for _ in range(n):
        result.append(a)
        a, b = b, a + b
    return result

fibonacci = fibonacci(10)
for num in fibonacci:
    print(num, end=' ')
    
````

````python
# write fibonacci series using recursion

def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)


for i in range(10):
    print(fibonacci(i), end=' ')
````

### 🧭 What is MRO (Method Resolution Order) in Python?

**MRO (Method Resolution Order)** is the **order in which Python looks for a method or attribute** when it is called on an object, particularly in **multiple inheritance** scenarios.

In simple terms:

> MRO determines **which method is executed first** when multiple classes define the same method name.

---

### 🧠 Why MRO Is Important

* **Avoids ambiguity** in multiple inheritance
* **Ensures consistent method lookup order**
* Prevents **diamond problem** (where multiple parent classes inherit from the same superclass)

---

### 🧪 How to Check MRO

You can check the MRO of a class using:

```python
print(ClassName.__mro__)
# or
ClassName.mro()
```

---

### 🔄 MRO Example with Multiple Inheritance

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

class C(A):
    def show(self):
        print("C")

class D(B, C):
    pass

obj = D()
obj.show()
print(D.mro())
```

#### 🟢 Output:

```
B
[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

### 📌 Explanation:

* Python looks for the method in `D` → then in `B` → then in `C` → then `A` → then `object`.
* It uses the **C3 linearization algorithm** to compute this order consistently.

---

### 🔧 Real-Time Use Case of MRO

#### ✅ 1. **Framework Design (e.g., Django, Flask)**

Frameworks often use **mixins** and multiple base classes to provide reusable functionality:

```python
class LoggingMixin:
    def process(self):
        print("Logging")

class ValidationMixin:
    def process(self):
        print("Validating")

class MyView(LoggingMixin, ValidationMixin):
    def process(self):
        super().process()
        print("Handling request")
```

MRO determines whether `LoggingMixin` or `ValidationMixin`'s `process()` runs first.

---

#### ✅ 2. **Diamond Problem Resolution**

```python
class A:
    def say(self):
        print("A")

class B(A):
    def say(self):
        print("B")

class C(A):
    def say(self):
        print("C")

class D(B, C):
    pass

D().say()
print(D.mro())
```

Even though both `B` and `C` inherit from `A`, Python resolves the method order using MRO to **avoid ambiguity**.

---

### ✅ Summary

| Concept   | Explanation                                                |
| --------- | ---------------------------------------------------------- |
| **MRO**   | Determines order in which classes are searched for methods |
| **Why**   | Helps manage method conflicts in multiple inheritance      |
| **How**   | Python uses **C3 Linearization** to build consistent order |
| **Tools** | Use `ClassName.mro()` or `ClassName.__mro__` to inspect    |

### ⚠️ Exception Handling in Python — Focus on `finally` Block

---

### 🔍 What is `finally`?

The `finally` block in Python is **used to define cleanup actions** that must be executed under all circumstances, **whether an exception occurs or not**.

It’s part of Python's exception handling structure:

```python
try:
    # risky code
except SomeException:
    # handle exception
finally:
    # this code runs no matter what
```

---

### ✅ Key Features of `finally`:

| Feature              | Description                                                            |
| -------------------- | ---------------------------------------------------------------------- |
| Guaranteed execution | Always runs, whether an exception was raised or not                    |
| Resource cleanup     | Commonly used to release resources (e.g., close file or DB connection) |
| Complements `try`    | Can be used with or without `except` block                             |

---

### 📌 Example 1: With Exception

```python
try:
    x = 1 / 0  # raises ZeroDivisionError
except ZeroDivisionError:
    print("Division by zero!")
finally:
    print("This always runs.")
```

**Output:**

```
Division by zero!
This always runs.
```

---

### 📌 Example 2: No Exception

```python
try:
    x = 10 / 2
    print("Result:", x)
finally:
    print("Cleanup done.")
```

**Output:**

```
Result: 5.0
Cleanup done.
```

---

### 📌 Example 3: Closing a File (Real-Time Use Case)

```python
try:
    file = open("data.txt", "r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found.")
finally:
    print("Closing file...")
    file.close()
```

✅ Even if an exception occurs, the file will be **closed** properly to avoid resource leaks.

---

### ⚠️ Real-Time Use Cases

| Scenario | Why Use `finally`?                             |
| -------- | ---------------------------------------------- |
| File I/O | Ensure file is closed after read/write         |
| Database | Ensure DB connection or cursor is closed       |
| Locking  | Release a threading or multiprocessing lock    |
| Logging  | Log exit/cleanup messages regardless of errors |

---

### ✅ Summary

| Block     | Purpose                             |
| --------- | ----------------------------------- |
| `try`     | Code that may throw an exception    |
| `except`  | Code to handle exceptions           |
| `finally` | Code that runs **always** (cleanup) |

## 2. **Tuple Challenge** — Partition by predicate

### Approach A: List Comprehension + Convert

```python
def partition_tuple(tup, pred):
    true_group = tuple(x for x in tup if pred(x))
    false_group = tuple(x for x in tup if not pred(x))
```

```python
def partition_tuple_one_pass(tup, pred):
    true, false = [], []
    for x in tup:
        (true if pred(x) else false).append(x)
    return tuple(true), tuple(false)
```

```python
def find_duplicates(lst):
    seen, dups = set(), set()
```

```python
def find_duplicates_counter(lst):
```

## 5. **Mixed Data Structures** — Rotate list by k

### Approach A: Slicing

```python
def rotate_slice(lst, k):
    k %= len(lst)
    return lst[-k:] + lst[:-k]

print(rotate_slice([1,2,3,4,5], 2))
# → [4, 5, 1, 2, 3]
```

### Approach B: Reverse In-place

```python
def reverse(lst, start, end):
    while start < end:
        lst[start], lst[end] = lst[end], lst[start]
        start += 1
        end -= 1

def rotate_inplace(lst, k):
    n = len(lst)
    k %= n
    reverse(lst, 0, n-1)
    reverse(lst, 0, k-1)
    reverse(lst, k, n-1)
    return lst

print(rotate_inplace([1,2,3,4,5], 2))
# → [4, 5, 1, 2, 3]
```

Here’s a step-by-step Python script that does this:

```python
import zipfile
import re
import os

zip_path = 'files.zip'        # Your ZIP file path
extract_dir = 'extracted'     # Folder to extract files

# 1. Extract all files
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# 2. Get list of extracted files
files = os.listdir(extract_dir)

# 3. Define a regex to capture the numeric part (e.g., '1.00.1')
pattern = re.compile(r'(\d+\.\d+\.\d+)_file\.txt')

# 4. Create a helper to convert the captured string to a tuple of ints for sorting
def sort_key(filename):
    match = pattern.match(filename)
    if match:
        parts = match.group(1).split('.')
        # Convert to int tuple: e.g., '1.00.3' -> (1, 0, 3)
        return tuple(int(p) for p in parts)
    return (float('inf'),)  # Push non-matching files to the end

# 5. Sort files using the helper
sorted_files = sorted(files, key=sort_key)

print("Sorted files:")
for f in sorted_files:
    print(f)
```

---

### Explanation:

* **Step 1:** Extract all files to the `extracted` folder.
* **Step 2:** List all extracted files.

# 🔶 OOP Concepts in Python

OOP is a programming paradigm based on the concept of **"objects"** — which bundle data and methods together. Python supports OOP naturally.

## 1. **Class & Object**

## 2. **Encapsulation**

* Hiding internal data and exposing only what is needed.


### ✅ Use Case:

* Reusing behavior in related classes like `Dog`, `Cat`, `Bird` inheriting `Animal`.

### ⚠️ Drawback:

* Deep inheritance chains can make code harder to read or debug (tight coupling).

## 4. **Polymorphism**

* Same method name, different behavior.
* Achieved via **method overriding** or **duck typing** in Python.

### ✅ Example:

```python

```

### ✅ Use Case:

* Writing flexible code that works with multiple object types (`Circle`, `Square`, etc.).

### ⚠️ Drawback:

* Can introduce bugs if expected interfaces are not implemented (Python is not statically typed by default).

### ✅ Use Case:

* Designing systems with interchangeable parts (e.g., different payment methods).

### ⚠️ Drawback:

* Requires additional code and structure, can be overkill for small projects.

## 🔹 4. **Class Method vs Static Method**

# 🧠 What is Multithreading?

**Multithreading** is a technique that allows **multiple threads** (smaller units of a process) to **run concurrently**.
Each thread runs independently, but shares memory with other threads of the same process.

---

## ✅ Why Use Multithreading?

* To make programs **more responsive**.
* To perform **I/O-bound** operations faster (e.g., file downloads, web scraping).
* **Not ideal** for CPU-bound tasks due to Python’s GIL (we'll explain this too).

---

## 🔧 Core Concepts

| Concept         | Description                                                                            |
| --------------- | -------------------------------------------------------------------------------------- |
| **Thread**      | A unit of execution within a process.                                                  |
| **Main thread** | The initial thread that runs when your script starts.                                  |
| **GIL**         | Global Interpreter Lock – allows only one thread to execute Python bytecode at a time. |
| **Thread-safe** | Code that behaves correctly when executed by multiple threads.                         |

---

## ✅ Example 4: Using ThreadPoolExecutor (Better)

```python
from concurrent.futures import ThreadPoolExecutor
import requests

urls = [
    'https://httpbin.org/image/png',
    'https://httpbin.org/image/jpeg',
    'https://httpbin.org/image/webp',

    response = requests.get(url)
    filename = url.split("/")[-1]
    with open(filename, 'wb') as f:

    return f"{filename} downloaded"

with ThreadPoolExecutor(max_workers=3) as executor:
    results = executor.map(download, urls)

for r in results:
    print(r)
```

## 🧱 The Problem: GIL (Global Interpreter Lock)

Python’s **GIL** ensures only **one thread** executes Python code at a time (in CPython).
This makes threading **inefficient for CPU-bound tasks** like:

* Large number crunching
* Data processing
* Image processing

✅ Use **`multiprocessing`** instead for CPU-bound tasks.

## 🔐 Thread Safety: Locks

```python

def increment():
    global counter
    for _ in range(100000):
        with lock:
            counter += 1

t1 = threading.Thread(target=increment)
t2 = threading.Thread(target=increment)

t1.start()
t2.start()
t1.join()
t2.join()

print("Counter:", counter)  # Should be 200000
```

# 🧠 What is Multiprocessing?

**Multiprocessing** is a technique that allows a program to **run multiple processes** simultaneously — taking advantage of **multiple CPU cores**.

Each process runs **independently**, has its **own memory**, and avoids the **Global Interpreter Lock (GIL)**.

---

## 🔄 Multiprocessing vs. Multithreading

| Feature           | Multithreading                    | Multiprocessing                        |
| ----------------- | --------------------------------- | -------------------------------------- |
| Memory            | Shared between threads            | Separate memory per process            |
| Speed (CPU-bound) | Slower due to GIL                 | Faster – uses multiple cores           |
| Best for          | I/O-bound tasks (e.g., file, web) | CPU-bound tasks (e.g., data crunching) |
| Crashes           | One thread crash = app lives      | One process crash = isolated           |

---

## ✅ When to Use Multiprocessing

* Heavy **data processing**
* **Parallel tasks** that don’t need to share memory
* Tasks that are CPU-intensive (e.g., image processing, ML, simulations)

---

## 🔧 Core Module: `multiprocessing`

Main tools:

* `Process` – run a function in a new process
* `Pool` – manage a pool of worker processes
* `Queue`, `Pipe` – for inter-process communication (IPC)
* `Lock` – to synchronize access to shared resources

---

## ✅ Example 1: Basic Process

```python
from multiprocessing import Process
import time

def task(name):
    print(f"Process {name} starting...")
    time.sleep(2)
    print(f"Process {name} done.")

if __name__ == '__main__':
    p1 = Process(target=task, args=("A",))
    p2 = Process(target=task, args=("B",))

    p1.start()
    p2.start()

    p1.join()
    p2.join()
    print("Both processes complete.")
```

## ✅ Example 2: Using `Pool` for Parallel Map

```python
from multiprocessing import Pool
import time

def square(n):
    time.sleep(1)
    return n * n

if __name__ == '__main__':
    numbers = [1, 2, 3, 4, 5]
    with Pool(processes=3) as pool:
        results = pool.map(square, numbers)
    print("Squares:", results)
```

### 🧠 What’s happening:

* `Pool` manages 3 processes.
* Tasks are divided and run **in parallel**.
* Total time ≈ `ceil(len(numbers) / num_processes) * task_time`

## ✅ Shared State (with `Value` and `Array`)

```python
from multiprocessing import Process, Value
import time

def increment(counter):
    for _ in range(100000):
        counter.value += 1

if __name__ == '__main__':
    count = Value('i', 0)
    p1 = Process(target=increment, args=(count,))

    p1.start()
    p2.start()
    p1.join()
    p2.join()
```

## ✅ Using a Lock to Avoid Race Conditions

```python
from multiprocessing import Process, Value, Lock

def safe_increment(counter, lock):
    for _ in range(100000):
        with lock:
            counter.value += 1

if __name__ == '__main__':
    counter = Value('i', 0)
    lock = Lock()

    p1 = Process(target=safe_increment, args=(counter, lock))
    p2 = Process(target=safe_increment, args=(counter, lock))

    p1.start()
    p2.start()
    p1.join()
    p2.join()

    print("Safe count:", counter.value)
```

## ✅ Interprocess Communication (IPC) with Queue

```python
from multiprocessing import Process, Queue

def produce(q):
    for i in range(5):
        q.put(i)
        print(f"Produced: {i}")

def consume(q):
    while not q.empty():
        item = q.get()
        print(f"Consumed: {item}")

if __name__ == '__main__':
    q = Queue()
    p1 = Process(target=produce, args=(q,))
    p1.start()
    p1.join()

    p2 = Process(target=consume, args=(q,))
    p2.start()
    p2.join()
```

# 🔄 What is `asyncio`?

`asyncio` is a built-in Python module that provides the infrastructure to write **concurrent code using the async/await syntax**.

---

## ✅ Key Difference from Threading and Multiprocessing

| Feature     | Threading       | Multiprocessing           | `asyncio`                               |
| ----------- | --------------- | ------------------------- | --------------------------------------- |
| Best for    | I/O-bound tasks | CPU-bound tasks           | I/O-bound tasks                         |
| Concurrency | OS threads      | OS processes              | Coroutine (lightweight tasks)           |
| Memory      | Shared          | Separate                  | Shared                                  |
| GIL Impact  | Blocked by GIL  | Avoids GIL with processes | Avoids GIL via cooperative multitasking |
| Overhead    | Moderate        | High                      | Very Low                                |

---

## 🔧 When to Use `asyncio`

* High-performance **web servers or APIs**
* **Web scraping** (non-blocking HTTP)
* **Chat apps**, **sockets**, **real-time systems**
* File I/O (with third-party libs like `aiofiles`)
* Avoids using multiple threads/processes for I/O

---

## 🔑 Key Concepts

| Concept         | Description                          |
| --------------- | ------------------------------------ |
| `async def`     | Defines an asynchronous coroutine    |
| `await`         | Waits for the result of a coroutine  |
| `asyncio.run()` | Starts the main coroutine            |
| `create_task()` | Run multiple coroutines concurrently |
| `sleep()`       | Non-blocking sleep                   |

---

## ✅ Basic Example: Async Sleep

```python
import asyncio

async def greet(name):
    print(f"Hello {name}")
    await asyncio.sleep(1)
    print(f"Goodbye {name}")

async def main():
    await asyncio.gather(
        greet("Alice"),
        greet("Bob"),
    )

asyncio.run(main())
```

⏱ Total time ≈ 1 second, **not 2**, because the coroutines run concurrently.

## ✅ Real-World Example: Web Requests

```python
import asyncio
import aiohttp  # pip install aiohttp

urls = [
    "https://httpbin.org/delay/1",
    "https://httpbin.org/delay/2",
    "https://httpbin.org/delay/3"
]

async def fetch(session, url):
    async with session.get(url) as response:
        print(f"Fetched {url}")
        return await response.text()

async def main():
    async with aiohttp.ClientSession() as session:
        tasks = [fetch(session, url) for url in urls]
        responses = await asyncio.gather(*tasks)
        print(f"Got {len(responses)} responses")

asyncio.run(main())
```

✅ **Faster than synchronous requests**, perfect for **scraping APIs**.

## ✅ Async Queue for Background Processing

```python
import asyncio

async def producer(queue):
    for i in range(5):
        print(f"Producing {i}")
        await queue.put(i)
        await asyncio.sleep(0.5)

async def consumer(queue):
    while True:
        item = await queue.get()
        print(f"Consumed {item}")
        queue.task_done()

async def main():
    queue = asyncio.Queue()
    producer_task = asyncio.create_task(producer(queue))
    consumer_task = asyncio.create_task(consumer(queue))

    await producer_task
    await queue.join()
    consumer_task.cancel()

asyncio.run(main())
```

✅ Great for **background job queues**, **pipeline systems**, **stream processing**.

# 🧵 `threading`

## ✅ What is it?

* Allows **concurrent execution** of tasks using multiple **threads** within a single process.
* Ideal for **I/O-bound** tasks (e.g., reading files, making API calls).

## ⚙️ How it works

* Multiple threads share the **same memory** space.
* Python threads are limited by the **Global Interpreter Lock (GIL)** – only one thread runs Python bytecode at a time.

## 🎯 When to use

* Tasks that **wait** a lot (e.g., network, disk I/O)
* When you don’t need heavy CPU usage

## 🔧 Real-World Use Cases

* Web crawlers
* File uploaders
* GUI background tasks
* Logging systems

## ⚠️ Limitations

* **Not suitable for CPU-bound tasks**
* Can cause **race conditions** — needs `Lock`, `Semaphore`, etc.

---

# 🧠 `multiprocessing`

## ✅ What is it?

* Runs **multiple processes** in parallel, each with **its own Python interpreter**.
* Great for **CPU-bound** tasks like image or data processing.

## ⚙️ How it works

* Each process has its **own memory space**.
* Can **truly parallelize** tasks across CPU cores.
* Avoids the GIL altogether.

## 🎯 When to use

* CPU-heavy operations
* Data crunching or transformations
* Running tasks in complete isolation

## 🔧 Real-World Use Cases

* Image processing (OpenCV, PIL)
* Data science (pandas, numpy calculations)
* ML model training
* Batch file conversions

## ⚠️ Limitations

* **High memory** consumption (copies data)
* Inter-process communication (IPC) is **harder/slower**
* More startup/shutdown overhead

---

# 🌀 `asyncio`

## ✅ What is it?

* A Python module for **asynchronous, cooperative multitasking**.
* Best for **I/O-bound** and **high concurrency** tasks (like thousands of HTTP requests).

## ⚙️ How it works

* Runs on a **single thread**, uses an **event loop**.
* Tasks yield control using `await` so others can run.

## 🎯 When to use

* High-concurrency apps (e.g., websockets, REST APIs)
* Thousands of small I/O tasks (network, file)

## 🔧 Real-World Use Cases

* Web scraping (aiohttp)
* Async file operations (aiofiles)
* Chat servers
* Streaming and notification systems

## ⚠️ Limitations

* **Not ideal for CPU-bound tasks**
* Doesn’t use multiple CPU cores
* Doesn’t preempt tasks — your coroutines must behave

# ⚙️ 1. What is an Iterator?

An **iterator** is an object that lets you **iterate over a sequence** one item at a time.

### ✅ Iterator Protocol

An object is an **iterator** if it:

* Implements `__iter__()` → returns itself
* Implements `__next__()` → returns next value or raises `StopIteration`

```python
class Counter:
    def __init__(self, limit):
        self.current = 0
        self.limit = limit

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.limit:
            val = self.current
            self.current += 1
            return val
        else:
            raise StopIteration

counter = Counter(3)
for num in counter:
    print(num)  # 0, 1, 2
```

---

# 🔄 2. `iter()` and `next()`

* `iter(obj)` calls `obj.__iter__()`
* `next(obj)` calls `obj.__next__()`

```python
nums = [10, 20, 30]
it = iter(nums)
print(next(it))  # 10
print(next(it))  # 20
print(next(it))  # 30
# next(it) would raise StopIteration
```

---

# ⚡ 3. Generator Functions: `yield`

A **generator** is a simpler way to create iterators using `yield`.

### ✅ Example:

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(3):
    print(num)  # 1, 2, 3
```

* `yield` **pauses** function state.
* Next call resumes **from where it left off**.

---

# 🔁 4. `yield from`: Delegate to sub-generator

Useful when breaking a generator into **smaller parts**.

```python
def sub():
    yield 1
    yield 2

def main():
    yield 0
    yield from sub()
    yield 3

for i in main():
    print(i)  # 0 1 2 3
```

---

# 🧠 5. Generator Expressions (One-liner)

Just like list comprehensions, but lazy:

```python
squares = (x*x for x in range(5))
print(next(squares))  # 0
print(next(squares))  # 1
```

✅ **Saves memory** compared to `[x*x for x in range(1_000_000)]`

# 📦 Real-World Use Case: Streaming Large Files

### Scenario:

Reading a **10GB log file** line-by-line without loading entire file in memory.

```python
def read_large_file(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()

for line in read_large_file("logs.txt"):
    if "ERROR" in line:
        print(line)
```

✅ **Memory-efficient**, **lazy-loaded**, fast

````python
'''
Withot using any libraries, we can implement the normalization manually.
This includes converting characters to lowercase, capitalizing the first letter of each word,
and ensuring that words are split and joined correctly. 
'''

# Convert a character to lowercase (ASCII-based)
def to_lower(char):
    if 'A' <= char <= 'Z':
        return chr(ord(char) + 32)
    return char

# Convert a character to uppercase (ASCII-based)
def to_upper(char):
    if 'a' <= char <= 'z':
        return chr(ord(char) - 32)
    return char

# Convert entire word to lowercase
def word_to_lower(word):
    result = ''
    for ch in word:
        result += to_lower(ch)
    return result

# Capitalize word: First letter uppercase, rest lowercase
def capitalize_word(word):
    if word == '':
        return ''
    first = to_upper(word[0])
    rest = ''
    for ch in word[1:]:
        rest += to_lower(ch)
    return first + rest

# Split sentence into words (manual split on space)
def split_words(text):
    words = []
    current = ''
    for ch in text:
        if ch == ' ':
            if current != '':
                words.append(current)
                current = ''
        else:
            current += ch
    if current != '':
        words.append(current)
    return words

# Join words with single space
def join_words(words):
    result = ''
    for i in range(len(words)):
        result += words[i]
        if i != len(words) - 1:
            result += ' '
    return result

# Final normalization function
def normalize_command(text):
    words = split_words(text)
    result_words = []
    for word in words:
        word_lower = word_to_lower(word)
        word_cap = capitalize_word(word_lower)
        result_words.append(word_cap)
    return join_words(result_words)

# Input
inputs = [
    'Navigate to starbucks',
    'navigate to sTarbUcks',
    'NAVIGATE to STARBUCKS'
]

# Output (with deduplication)
output = []
seen = []

for phrase in inputs:
    normalized = normalize_command(phrase)
    if normalized not in seen:
        seen.append(normalized)
        output.append(normalized)

# Print result
for line in output:
    print(line)
````

### ✅ Example: Multiple Inheritance with `super()`

We'll demonstrate how `super()` works **with multiple parent classes**, using `super()` to call parent methods in the correct MRO (Method Resolution Order).

```python
class A:
    def show(self):
        print("This is A")
        super().show()

class B:
    def show(self):
        print("This is B")
        super().show()

class C:
    def show(self):
        print("This is C")

class D(A, B, C):
    def show(self):
        print("This is D")
        super().show()

# Create object of D
obj = D()
obj.show()

# Show MRO
print(D.__mro__)
```

---

### ✅ Output:

```
This is D
This is A
This is B
This is C
(<class '__main__.D'>, <class '__main__.A'>, <class '__main__.B'>, <class '__main__.C'>, <class 'object'>)
```

---

### 🔍 Explanation:

* `D` inherits from `A`, `B`, and `C`.

* Each class uses `super().show()` to call the next method in the MRO chain.

* `super()` respects the **MRO**, so the order of method calls is:

  ```
  D → A → B → C
  ```

* `C` is the last one and doesn't call `super()`, so it stops there.

---

### 🧠 Key Notes on `super()` and Multiple Inheritance:

* `super()` is **not limited to the direct parent** — it moves to the **next class in the MRO**.
* Use `super()` consistently to make cooperative multiple inheritance work correctly.
* Always check MRO with `ClassName.__mro__` if you're unsure of method resolution.

## The Problem We’re Solving

You have multiple classes in a hierarchy, and you want each class to have a `show()` method that:

* Prints something like `"This is <ClassName> class"`.
* Calls the parent class’s `show()` method automatically (if it exists), so all messages from the class hierarchy get printed in order.

You want to **avoid writing the `show()` method in every class manually**, so we use a **decorator** to add it automatically.

---

## The Decorator Code

```python
def add_show(cls):
    def show(self):
        print(f"This is {cls.__name__} class")  # Print the current class name
        
        # Call the next class's show() in the MRO chain if it exists
        try:
            super(cls, self).show()
        except AttributeError:
            pass  # No parent show() method to call (base case)

    cls.show = show  # Inject this show method into the class
    return cls       # Return the modified class
```

---

### What is going on here?

1. **`def add_show(cls):`**

   This is the decorator function — it takes a class `cls` as input.

2. **`def show(self):`**

   Defines a new method `show()` to add to the class dynamically.

3. **`print(f"This is {cls.__name__} class")`**

   Prints the message using the class name (`cls.__name__`).

4. **`super(cls, self).show()`**

   * This is a key line.
   * It calls the **next class’s** `show()` method in the MRO *after* `cls`.
   * Normally, `super()` is used inside methods, but here we specify `cls` explicitly to tell Python:
     "Find the next class after `cls` in the MRO for this instance `self`."
   * This enables chaining all `show()` methods in the hierarchy, respecting MRO order.

5. **`except AttributeError: pass`**

   If no `show()` method exists in the next class (usually the last class, like `object`), just ignore the error and stop calling further.

6. **`cls.show = show`**

   Dynamically attaches this new `show` method to the class `cls`.

7. **`return cls`**

   Returns the modified class back, so it replaces the original class definition.

---

## How You Use It

```python
@add_show
class A:
    def __init__(self):
        print("Init A")
```

* When Python reads this, it calls `add_show(A)`.
* The decorator adds the `show()` method to `A` automatically.
* You do the same for `B`, `C`, and `D`.

---

## What Happens When You Call `obj.show()` on `D`?

1. The `show()` method added by the decorator for `D` runs:

   * Prints `"This is D class"`
   * Calls `super(D, obj).show()` → calls `B`’s `show()` (next in MRO).

2. `B`’s decorator-added `show()` runs:

   * Prints `"This is B class"`
   * Calls `super(B, obj).show()` → calls `C`’s `show()`.

3. `C`’s decorator-added `show()` runs:

   * Prints `"This is C class"`
   * Calls `super(C, obj).show()` → calls `A`’s `show()`.

4. `A`’s decorator-added `show()` runs:

   * Prints `"This is A class"`
   * Calls `super(A, obj).show()` → no more `show()`; catches `AttributeError` and stops.

---

## Summary

* **Decorator `add_show` adds the same kind of `show()` to every class automatically.**
* **Each `show()` method prints its own class’s message, then calls the next in MRO using `super(cls, self).show()`.**
* **This creates a chain of `show()` calls in the proper order, without you writing the method manually each time.**

````python
'''✅ 1. Two Sum – (Hashing, Arrays)
🔍 Problem:
Find two numbers in an array that add up to a given target.

💡 Logic:
Use a dictionary (hashmap) to store numbers and their indexes.

For each number, calculate the difference needed to reach the target.

If that difference is already in the hashmap, you've found the two numbers.'''

def two_sum(nums, target):
    hashmap = {}  # To store numbers and their indices
    for i, num in enumerate(nums):
        complement = target - num  # What number we need to add to get the target
        if complement in hashmap:
            return [hashmap[complement], i]  # Found it
        hashmap[num] = i  # Store number and its index
    return []

two_sum([2, 7, 11, 15], 9)
# Steps:
# i=0, num=2 → needs 7 → not found
# i=1, num=7 → needs 2 → found in hashmap at index 0
# ✅ Answer: [0, 1]
````

````python

'''✅ 3. Find First and Last Position of Element in Sorted Array – (Binary Search)
📘 Problem Statement:
Given a sorted array of integers nums, and a target, return the starting and ending index of the target value.

If the target is not found, return [-1, -1].

🧠 Intuition:
Since the array is sorted, use binary search to find:

The first index where the target appears.

The last index where the target appears.

🧪 Example:
python
Copy
Edit
nums = [5,7,7,8,8,10]
target = 8
8 appears at index 3 and 4

✅ Output: [3, 4]'''

def find_first_and_last(nums, target):
    def find_bound(is_first):
        low, high = 0, len(nums) - 1
        result = -1

        while low <= high:
            mid = (low + high) // 2
            if nums[mid] == target:
                result = mid  # Save match
                if is_first:
                    high = mid - 1  # Search left
                else:
                    low = mid + 1   # Search right
            elif nums[mid] < target:
                low = mid + 1
            else:
                high = mid - 1
        return result

    return [find_bound(True), find_bound(False)]

# Example usage
nums = [5, 7, 7, 8, 8, 10]
target = 8
result = find_first_and_last(nums, target)
print(result)  
````

## ✅ 1. **What is a regular expression?**

**Answer:**
A regular expression (regex) is a sequence of characters that defines a search pattern. It's used for pattern matching within strings—like searching, extracting, replacing, or validating data such as email addresses, phone numbers, and more.

---

## ✅ 2. **How is regex different from simple string functions like `find()` or `split()`?**

**Answer:**

* Regex offers **pattern-based matching**, while functions like `find()` or `split()` work on exact strings or simple delimiters.
* Regex is much more **flexible and powerful** for complex patterns (like extracting dates, validating formats, etc.).
* For example, to extract all dates or phone numbers from text, regex is far superior.

---

## ✅ 3. **What are the basic components of regex syntax?**

**Answer:**

* **Literals**: Regular characters (e.g., `a`, `1`)
* **Meta-characters**: Special symbols (`.`, `^`, `$`, `*`, `+`, `?`, `{}`, `[]`, `()`, `|`, `\`)
* **Character classes**: `[abc]`, `\d`, `\w`, etc.
* **Quantifiers**: `*`, `+`, `?`, `{n}`, `{n,m}`
* **Anchors**: `^` (start), `$` (end)
* **Groups**: `()`, `(?:)` for non-capturing
* **Alternation**: `|` (OR)

---

## ✅ 4. **What is the difference between `re.match()`, `re.search()`, and `re.findall()` in Python?**

**Answer:**

* `re.match()` checks **only at the beginning** of the string.
* `re.search()` scans the **entire string** for the first match.
* `re.findall()` returns **all non-overlapping matches** in a list.

---

## ✅ 5. **What are greedy and non-greedy (lazy) quantifiers?**

**Answer:**

* **Greedy quantifiers** (`*`, `+`, `{n,m}`): Match as much text as possible.
* **Non-greedy (lazy) quantifiers** (`*?`, `+?`, `{n,m}?`): Match as little as possible.

**Example:**

```python
import re
re.findall(r'<.*>', '<b>Bold</b><i>Italic</i>')   # Greedy → ['<b>Bold</b><i>Italic</i>']
re.findall(r'<.*?>', '<b>Bold</b><i>Italic</i>')  # Non-greedy → ['<b>', '</b>', '<i>', '</i>']
```

---

## ✅ 6. **What are character classes and how do they work?**

**Answer:**
Character classes let you match one of several characters:

* `[aeiou]` matches any vowel
* `[0-9]` matches any digit
* `[^abc]` matches anything except `a`, `b`, or `c`
* Shorthand: `\d` (digits), `\w` (word chars), `\s` (whitespace)

---

## ✅ 7. **What does the `^` and `$` symbol mean in regex?**

**Answer:**

* `^` asserts the **start** of a string.
* `$` asserts the **end** of a string.

**Example:**

* `^abc` matches "abc" only if it's at the beginning.
* `abc$` matches "abc" only if it's at the end.

---

## ✅ 8. **What is the use of `()` in regex?**

**Answer:**
Parentheses `()` create **capture groups**, which:

* Extract matched substrings
* Apply quantifiers to grouped patterns
* Help with back-references

**Example:**

```python
match = re.search(r'(\d{4})-(\d{2})-(\d{2})', '2025-07-15')
print(match.group(1))  # 2025
```

---

## ✅ 9. **What are lookahead and lookbehind assertions?**

**Answer:**

* **Lookahead** (`(?=...)`): Matches only if the next part of the string matches the pattern.
* **Lookbehind** (`(?<=...)`): Matches only if the previous part matches.

**Example:**

```python
re.findall(r'\d+(?=%)', '50% of 200 is 100')   # → ['50'] (match number before %)
re.findall(r'(?<=\$)\d+', '$100 and $200')    # → ['100', '200'] (match number after $)
```

---

## ✅ 10. **What is a backreference in regex?**

**Answer:**
A backreference (like `\1`) lets you refer to a previously matched group.

**Example:**

```python
re.findall(r'(\b\w+)\s+\1\b', 'hello hello world')  # → ['hello']
```

This detects repeated words using a backreference to the first group.

```regex
^[\w\.-]+@[\w\.-]+\.\w{2,}$
```

Matches basic emails like `test.user@example.com`.

| `+`        | 1 or more | `a+` → 'a', 'aaa'     |

````python
def flatten(nested_list):
    flat_list = []
    for item in nested_list:
        if isinstance(item, list):
            flat_list.extend(flatten(item))  # Recursively flatten sublist
        else:
            flat_list.append(item)
    return flat_list

# Example with deep nesting
nested = [1, [2, [3, 4], 5], 6, [[7, 8], 9]]

flattened = flatten(nested)
print("Flattened list:", flattened)
````

````python
def flatten_generator(nested):
    for item in nested:
        if isinstance(item, (list, tuple)):
            # Recursively yield items from sublist or tuple
            yield from flatten_generator(item)
        else:
            yield item

# Example with nested lists and tuples
nested = [1, (2, [3, 4], 5), 6, [[7, 8], 9]]

flat_gen = flatten_generator(nested)
flat_list = list(flat_gen)
print("Flattened list using generator:", flat_list)
````

## 🧩 What is Dependency Injection?

**Dependency Injection (DI)** is a design pattern where **objects or functions get the dependencies they need from an external source** rather than creating them internally.

> 📌 In FastAPI, `Depends()` is used to **inject dependencies** like database sessions, authentication, reusable logic, etc.

* Promotes **separation of concerns**
* Makes code more **modular and reusable**
* Simplifies **testing**
* Centralizes shared logic (like DB, auth, configs)

## 🧪 Example 1: Simple Dependency

## 🔐 What is JWT Authentication?

**JWT** stands for **JSON Web Token**. It's a compact, URL-safe token used to **securely transmit information** between parties as a **JSON object**.

It's often used in APIs to authenticate users — like a digital passport or ID card.

---

## 🧩 How JWT Authentication Works

Here's the flow:

1. **User Logs In**:

   * User sends their **username and password** to your API.
   * If correct, the server returns a **JWT token**.

2. **Client Stores Token**:

   * The client (browser, mobile app) stores this token (usually in localStorage or memory).

3. **Client Sends Token on Requests**:

   * For every request to a protected route, the client sends the token in the **Authorization header**:

     ```
     Authorization: Bearer <JWT_TOKEN>
     ```

4. **Server Verifies Token**:

   * The server **decodes** and **verifies** the token using a **secret key**.
   * If valid, the request proceeds.
   * If not, it returns a 401 (Unauthorized).

---

## 🧪 Example JWT Token (decoded)

```json
Header:
{
  "alg": "HS256",
  "typ": "JWT"
}

Payload:
{
  "user_id": 123,
  "exp": 1721498500  # Expiration timestamp
}

Signature:
HMACSHA256(
  base64UrlEncode(header) + "." + base64UrlEncode(payload),
  secret_key
)
```

Here’s a basic setup:

### 🔹 1. Install `python-jose` for JWT handling:

### 🔹 2. Create and verify JWT tokens

```python
from fastapi import FastAPI, Depends, HTTPException
from jose import JWTError, jwt
from datetime import datetime, timedelta

SECRET_KEY = "mysecretkey"
ALGORITHM = "HS256"
ACCESS_TOKEN_EXPIRE_MINUTES = 30

app = FastAPI()

# Create token
def create_access_token(data: dict):
    to_encode = data.copy()
    expire = datetime.utcnow() + timedelta(minutes=ACCESS_TOKEN_EXPIRE_MINUTES)
    to_encode.update({"exp": expire})
    return jwt.encode(to_encode, SECRET_KEY, algorithm=ALGORITHM)

# Dependency to verify token
def get_current_user(token: str = Depends(lambda: "fake-token-from-header")):
    try:
        payload = jwt.decode(token, SECRET_KEY, algorithms=[ALGORITHM])
        user_id: str = payload.get("user_id")
        if user_id is None:
            raise HTTPException(status_code=401, detail="Invalid token")
        return {"user_id": user_id}
    except JWTError:
        raise HTTPException(status_code=401, detail="Invalid token")

# Login route
@app.post("/login")
def login():
    # Normally you'd verify username/password here
    access_token = create_access_token(data={"user_id": 123})
    return {"access_token": access_token, "token_type": "bearer"}

# Secure route
@app.get("/profile")
def read_profile(user: dict = Depends(get_current_user)):
    return {"user": user}
```

> 🔐 Replace `lambda: "fake-token-from-header"` with actual logic to extract from headers in real use.

````python
# FIND A PRIME NUMBER
def is_prime(num):
    if num <= 1:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(num**0.5) + 1):  # Check divisibility up to the square root of num
        if num % i == 0:
            return False  # Found a divisor, so it's not prime
    return True  # No divisors found, so it's prime

# Example usage
number = 29
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
````

````python
#check strings are anagrams or not without using sorted
def are_anagrams(str1, str2):
    ''' Check if two strings are anagrams of each other. '''
    # explain below code    
    # The function first removes spaces and converts both strings to lowercase.
    # It then checks if the lengths of the two strings are equal.   
    # If they are not equal, it returns False immediately.  
    # If they are equal, it counts the occurrences of each character in the first string using a dictionary.
    # It then checks each character in the second string against this count.    
    # If a character is found in the second string, it decrements its count.
    # If the count goes below zero or if a character is not found in the dictionary,
    # it returns False. Finally, it checks if all counts in the dictionary are zero, indicating that the strings are anagrams.

    
    # Remove spaces and convert to lowercase
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    if len(str1) != len(str2):
        return False  # Different lengths cannot be anagrams

    char_count = {}

    for char in str1:
        char_count[char] = char_count.get(char, 0) + 1

    for char in str2:
        if char in char_count:
            char_count[char] -= 1
            if char_count[char] < 0:
                return False
        else:
            return False
    print(char_count)
    # Finally, check if all counts are zero 
    return all(count == 0 for count in char_count.values())         

# Example usage
str1 = 'listen'    
str2 = 'silent'
if are_anagrams(str1, str2):
    print(f'"{str1}" and "{str2}" are anagrams.')   
else:
    print(f'"{str1}" and "{str2}" are not anagrams.')
````

````python
#check  if string is palindrome or not without using slicing
def is_palindrome(s):
    s = s.lower()  # Convert to lowercase for case-insensitive comparison
    left, right = 0, len(s) - 1  # Initialize pointers
    while left < right:
        # example: left = 0, right = 3 for "abba"
        if s[left] != s[right]:  # Compare characters
            return False  # Not a palindrome
        left += 1  # Move left pointer forward
        right -= 1  # Move right pointer backward
    return True  # All characters matched   
# Example usage
input_string = "abba"
if is_palindrome(input_string):
    print(f'"{input_string}" is a palindrome.')
else:
    print(f'"{input_string}" is not a palindrome.')
````

````python
# factorial of a number using recursion
def factorial(n):
    ''' Calculate factorial of a number using recursion. '''
    if n == 0 or n == 1:  # Base case: factorial of 0 or 1 is 1
        return 1
    else:
        return n * factorial(n - 1)  # Recursive case: n * factorial of (n-1)
    
factorial_number = 5
result = factorial(factorial_number)
print(f"Factorial of {factorial_number} is {result}")
````

````python
# factorial of a number without using recursion
def factorial(n):
    ''' Calculate factorial of a number without recursion. '''
    result = 1
    for i in range(2, n + 1):  # Start from 2 to n
        result *= i  # Multiply result by current number
        print(f"Intermediate result after multiplying by {i}: {result}")  # Debugging line to show intermediate results
    return result
factorial_number = 7
result = factorial(factorial_number)

print(f"Factorial of {factorial_number} is {result}")
````

````python
# factorial of a number without recursion and without using a loop , using generator
def factorial(n):
    ''' Calculate factorial of a number using a generator. '''
    def generate_factorial():
        result = 1
        for i in range(2, n + 1):
            result *= i
            yield result  # Yield the intermediate result

    gen = generate_factorial()
    final_result = None
    for value in gen:
        final_result = value  # Get the last yielded value as the final result
    return final_result

factorial_number = 6
result = factorial(factorial_number)
print(f"Factorial of {factorial_number} is {result}")
````

````python
# find length of a array without using len() function
def find_length(arr):
    ''' Find the length of an array without using len() function. '''
    count = 0
    for _ in arr:  # Iterate through each element in the array
        count += 1  # Increment count for each element
    return count
# Example usage
arr = [1, 2, 3, 4, 5]
length = find_length(arr)
print(f"Length of the array is: {length}")
````

````python
# find length of a array without using len() function but using recursion
def find_length(arr):
    ''' Find the length of an array using recursion. '''
    if not arr:  # Base case: if the array is empty
        return 0
    else:
        print(f"Current element: {arr[0]}")
        # Recursive case: count current element + length of the rest
        print(f"Remaining array: {arr[1:]}")
        # Recursively call find_length on the rest of the array
        return 1 + find_length(arr[1:])  # Recursive case: count current element + length of the rest
    
# Example usage
arr = [1, 2, 3, 4, 5]
length = find_length(arr)
print(f"Length of the array is: {length}")
````

````python
# lenght of an array with generator
def find_length(arr):
    ''' Find the length of an array using a generator. '''
    def length_generator():
        for _ in arr:  # Iterate through each element in the array
            yield 1  # Yield 1 for each element

    gen = length_generator()
    total_length = sum(gen)  # Sum all yielded values to get the total length
    return total_length

# Example usage
arr = [1, 2, 3, 4, 5]
length = find_length(arr)
print(f"Length of the array is: {length}")
````

````python
# find a max marks for list of tuples
def find_max_marks(students):
    ''' Find the maximum marks from a list of tuples (name, marks). '''
    if not students:  # Check if the list is empty
        return None  # Return None if no students are present

    max_marks = students[0][1]  # Initialize with the first student's marks
    max_student = students[0][0]  # Initialize with the first student's name

    for student in students:
        print(f"Checking student: {student[0]}, Marks: {student[1]}")  # Debugging line to show current student
        if student[1] > max_marks:  # Compare current student's marks with max_marks
            max_marks = student[1]  # Update max_marks
            max_student = student[0]  # Update max_student

    return max_student, max_marks  # Return the student with maximum marks and their marks

# Example usage

students = [("Alice", 85), ("Bob", 92), ("Charlie", 78), ("David", 95)]
max_student, max_marks = find_max_marks(students)
print(f"Student with maximum marks: {max_student}, Marks: {max_marks}")
````

````python
# input is aaabccddd output is a3b1c2d3
def compress_string(s):
    ''' Compress a string by counting consecutive characters. '''
    if not s:  # Check if the string is empty
        return ""
    
    compressed = []  # List to store compressed parts
    count = 1  # Initialize count for the first character
    
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:  # If current character is same as previous
            count += 1  # Increment count
        else:
            compressed.append(f"{s[i - 1]}{count}")  # Append character and count to the list
            count = 1  # Reset count for the new character
    
    compressed.append(f"{s[-1]}{count}")  # Append the last character and its count
    print("Compressed parts:", compressed)  # Debugging line to show compressed parts
    return ''.join(compressed)  # Join the list into a string
# Example usage
input_string = "aaabccddd"  
result = compress_string(input_string)
print(f"Compressed string of '{input_string}': {result}")
````

````python
# input is a3b1c2d3 output is aaaabccddd
def decompress_string(s):
    ''' Decompress a string by expanding characters based on their counts. '''
    decompressed = []  # List to store decompressed parts
    i = 0  # Initialize index
    
    while i < len(s):
        char = s[i]  # Get the character
        i += 1  # Move to the next character
        count = 0  # Initialize count
        
        # Read the number that follows the character
        while i < len(s) and s[i].isdigit():
            print(f"Current character: {char}, Current index: {i}, Current count: {count}")
            count = count * 10 + int(s[i])  # Build the count from digits
            print(f"Current count for '{char}': {count}")
            i += 1  # Move to the next character
        
        decompressed.append(char * count)  # Append the character repeated 'count' times
        print(f"Decompressed part: {char * count}")
    return ''.join(decompressed)  # Join the list into a string

# Example usage
input_string = "a3b1c2d3"
result = decompress_string(input_string)
print(f"Decompressed string of '{input_string}': {result}")
````

````python
# defination of decorator , its uses and explain how it works
# A decorator in Python is a special type of function that can modify or enhance the behavior of another function or method.
# It allows you to wrap another function, providing a way to add functionality before or after the
# original function is called, without modifying the original function's code.
# Decorators are often used for logging, access control, memoization, and other cross-cutting concerns.
# Decorators are defined using the `@decorator_name` syntax above the function definition.
def my_decorator(func):
    ''' A simple decorator that prints a message before and after the function call. '''
    def wrapper(*args, **kwargs):
        print("Before calling the function.")
        result = func(*args, **kwargs)  # Call the original function
        print("After calling the function.")
        return result  # Return the result of the original function
    return wrapper  # Return the wrapper function
# Example usage of the decorator
@my_decorator
def say_hello(name):
    ''' A simple function that prints a greeting. '''
    print(f"Hello, {name}!")
# Calling the decorated function
say_hello("Alice")
# The output will show the messages from the decorator before and after the greeting.
# Decorators can also take arguments, allowing for more flexible behavior.
def repeat_decorator(times):
    ''' A decorator that repeats the function call a specified number of times. '''
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(times):  # Repeat the function call 'times' times
                func(*args, **kwargs)  # Call the original function
        return wrapper  # Return the wrapper function
    return decorator  # Return the decorator function
# Example usage of the repeat decorator
@repeat_decorator(3)    
def greet(name):
    ''' A simple function that prints a greeting. '''
    print(f"Hello, {name}!")
# Calling the decorated function
greet("Bob")
# The output will show the greeting printed three times due to the decorator.
# Decorators can also be used with classes, allowing you to modify class methods or properties.
def class_decorator(cls):
    ''' A decorator that adds a new method to the class. '''
    cls.new_method = lambda self: print("This is a new method added by the decorator.")
    return cls  # Return the modified class
# Example usage of the class decorator
@class_decorator
class MyClass:
    ''' A simple class. '''
    def original_method(self):
        print("This is the original method.")

# Creating an instance of the decorated class
my_instance = MyClass()
# Calling the original method
my_instance.original_method()
# Calling the new method added by the decorator
my_instance.new_method()
# Decorators can also be used to modify the behavior of class methods.
def method_decorator(method):
    ''' A decorator that modifies the behavior of a class method. '''
    def wrapper(self, *args, **kwargs):
        print("Before calling the method.")
        result = method(self, *args, **kwargs)  # Call the original method
        print("After calling the method.")
        return result  # Return the result of the original method
    return wrapper  # Return the wrapper function
# Example usage of the method decorator
class MyClass:
    ''' A simple class with a method. '''
    @method_decorator
    def my_method(self, name):
        print(f"Hello, {name}!")
# Creating an instance of the class
my_instance = MyClass()
# Calling the decorated method
my_instance.my_method("Charlie")
# The output will show the messages from the method decorator before and after the greeting.
````

**Monkey patching** is a technique in Python where you **dynamically modify or extend code at runtime**, especially methods or attributes of classes, modules, or even built-in types.

---

### 🔧 Definition:

> **Monkey patching** means changing or replacing parts of code **without altering the original source code**, usually to fix bugs, add features, or modify behavior during execution.

---

### 🐒 Why “monkey”?

It's a playful term implying that you're "hacking" or "tweaking" existing code—often in a non-standard or risky way—like a monkey messing with a machine.

---

### ✅ Example – Monkey Patching a Class Method:

```python
class Dog:
    def bark(self):
        print("Woof!")

# Original behavior
dog = Dog()
dog.bark()  # Output: Woof!

# Monkey patch the bark method
def new_bark(self):
    print("I'm a quiet dog.")

Dog.bark = new_bark

# New behavior
dog.bark()  # Output: I'm a quiet dog.
```

---

### ✅ Example – Patching a Standard Library Function:

```python
import time

# Monkey patch time.sleep
def fast_sleep(seconds):
    print(f"Skipping sleep of {seconds} seconds")

time.sleep = fast_sleep

time.sleep(5)  # Output: Skipping sleep of 5 seconds
```

This is often used in **testing** to avoid delays or simulate behavior.

---

---

## 🔸 1. **Private Heap Space**

* All Python objects and data structures (like lists, dictionaries, numbers, etc.) are stored in a **private heap**.
* This heap is **managed by the Python memory manager**, and programmers don’t have direct access to it.

> Think of it as a special storage area just for Python to allocate and organize memory for its objects.

---

## 🔸 2. **Automatic Memory Management**

* Python automatically **allocates memory** when you create an object and **frees it** when it's no longer used.
* You don’t have to use `malloc` or `free` like in C or C++.
* This process is handled by the **Python memory manager**, which works with the garbage collector.

---

## 🔸 3. **Reference Counting**

Python keeps track of how many **references** point to an object. Each object has an internal counter:

```python
import sys

a = []            # Create a list
b = a             # Another reference to the same list
print(sys.getrefcount(a))  # Usually returns 3 (a, b, and the argument to getrefcount)
```

* When the reference count of an object becomes **zero**, the memory is **released**.
* This is the **main mechanism** for memory management.

---

```python
a.ref = b
b.ref = a  # Circular reference!
```

* Here, both `a` and `b` refer to each other, so their reference count never drops to zero.
* Python’s **garbage collector** detects such **cyclic references** and cleans them up.

* Python uses a specialized allocator called **`pymalloc`** to manage memory efficiently.
* Instead of requesting memory from the OS each time, it uses **memory pools** for small objects (<= 512 bytes).
* This **improves performance** and **reduces fragmentation**.

## 📌 Key Difference: Shallow Copy vs Deep Copy

| Feature           | **Shallow Copy**                  | **Deep Copy**                                |
| ----------------- | --------------------------------- | -------------------------------------------- |

## 🔁 Shallow Copy

* Creates a **new object**, but inserts **references** to the **same inner objects**.

### 🧪 Example:

```python

shallow[0][0] = 99

```

👉 Both `original` and `shallow` share the same inner lists. Changing one affects the other.

---

## 🔁 Deep Copy

* Creates a **new object**, and **recursively copies all nested objects**.
* No shared references — completely independent.

### 🧪 Example:

```python
import copy

original = [[1, 2], [3, 4]]
deep = copy.deepcopy(original)

deep[0][0] = 99
print("Original:", original)  # [[1, 2], [3, 4]]
print("Deep:", deep)          # [[99, 2], [3, 4]]
```

👉 Now they’re independent — changing `deep` does **not** affect `original`.

> So assignment copies the **reference**, not the object.

## 🔄 Immutable Example

```python

y = x
x = 20
print(y)  # Output: 10
```

* `int` is immutable, so when you do `x = 20`, it creates a **new int object** (`20`), and `x` points to it.
* `y` still points to the original object (`10`).

```python

list2 = list1
list1.append(3)

print(list2)  # Output: [1, 2, 3]
```

* `list2` references the same object as `list1`.
* Mutating `list1` affects `list2` as well.

## 🔍 What is Scope in Python?

A **scope** defines where a variable is accessible in your code.

When you reference a variable, Python **resolves it by looking in a specific order**, called the **LEGB rule**.

---

## 🧠 LEGB Rule — Python’s Scope Resolution Order

LEGB stands for:

| Scope Type        | Meaning                                     | Example                         |
| ----------------- | ------------------------------------------- | ------------------------------- |
| **L - Local**     | Inside the current function                 | Variables defined in a function |
| **E - Enclosing** | Inside enclosing (outer) functions          | Variables in a nested function  |
| **G - Global**    | At the top-level of the module/script       | Variables outside any function  |
| **B - Built-in**  | Python’s built-in names (like `len`, `str`) | From `__builtins__` module      |

---

## 🔁 How Scope Resolution Works

Python searches for a variable in this order:

1. **Local**
2. **Enclosing**
3. **Global**
4. **Built-in**

It uses the **first match** it finds.

---

### 🧪 Example:

```python
x = "global"

def outer():
    x = "enclosing"
    
    def inner():
        x = "local"
        print(x)  # Looks for x using LEGB
    
    inner()

outer()
```

**Output:** `local` — because `x` was found in the **local scope** of `inner()`.

---

### 🔁 Now remove `x = "local"`:

```python
x = "global"

def outer():
    x = "enclosing"
    
    def inner():
        print(x)  # Now Python goes to Enclosing
    inner()

outer()
```

**Output:** `enclosing`

---

## 🔨 Global and Nonlocal Keywords

### 📌 `global`: tells Python to use the global variable

```python
x = 10

def modify():
    global x
    x = 20

modify()
print(x)  # 20
```

---

### 📌 `nonlocal`: tells Python to use a variable from the **enclosing** scope (not global)

```python
def outer():
    x = "outer"
    def inner():
        nonlocal x
        x = "changed"
    inner()
    print(x)

outer()  # Output: changed
```

---

## 📝 Summary: LEGB Scope Chain

| Order | Name          | Scope Description                            |
| ----- | ------------- | -------------------------------------------- |
| 1     | **Local**     | Inside the current function                  |
| 2     | **Enclosing** | Any outer function(s), in a nested structure |
| 3     | **Global**    | Top-level of the current module              |
| 4     | **Built-in**  | Predefined in Python (e.g., `len`, `range`)  |

---

## ⚠️ Common Mistakes

* Modifying a variable without declaring `global` or `nonlocal` when needed.
* Assuming a nested function can directly modify outer variables (you need `nonlocal`).

---

## 📌 Summary Table

| Type         | Where It Lives                  | Accessed In              | Modified With Keyword |
| ------------ | ------------------------------- | ------------------------ | --------------------- |
| **Local**    | Inside the current function     | That function only       | (default)             |
| **Global**   | At the top level of a module    | Anywhere (with `global`) | `global`              |
| **Nonlocal** | Enclosing function (not global) | Nested functions only    | `nonlocal`            |

---

## 🧠 1. Local Variables

* **Defined inside a function.**
* Exists only **within that function**.
* Automatically created when assigned inside the function.

```python
def func():
    x = 10  # Local variable
    print(x)

func()  # Output: 10
```

---

## 🌍 2. Global Variables

* **Defined at the top level** of a script/module.
* Accessible from any function **if not shadowed by a local variable**.
* To **modify** it inside a function, you must use the `global` keyword.

```python
x = "global"

def change():
    global x
    x = "modified"

change()
print(x)  # Output: modified
```

Without `global`, the function would just create a new **local `x`**.

---

## 🔁 3. Nonlocal Variables

* Used inside **nested functions**.
* Refers to a variable in the **nearest enclosing (non-global) scope**.
* Allows modification of a variable in the **outer function**, without making it global.

```python
def outer():
    x = "outer"
    def inner():
        nonlocal x
        x = "modified"
    inner()
    print(x)

outer()  # Output: modified
```

Without `nonlocal`, `x = "modified"` would create a **local `x`** in `inner()`.

---

## 🚫 What if You Don’t Use `global` or `nonlocal`?

Python will **treat the variable as local** and **not modify the outer one**, leading to bugs:

```python
x = 5

def broken():
    x = x + 1  # ❌ UnboundLocalError!

broken()
```

Why? Because Python sees `x = ...` and assumes it's local, but it's being read **before it's assigned** locally.

---

## ✅ When to Use

| Scenario                              | Use        |
| ------------------------------------- | ---------- |
| Just working inside one function      | `local`    |
| Sharing state across entire module    | `global`   |
| Sharing state across nested functions | `nonlocal` |

```python
        print("Scanning")
```

If a machine only prints, it doesn’t need to scan too!

## 🧠 **Simple Explanation:**

**Metaclasses** are the **classes of classes**.

* In Python, everything is an object — **even classes themselves are objects**.
* Metaclasses are what **create** those class objects.

> ✅ If **objects are created by classes**, then **classes are created by metaclasses**.

---

## 🧱 **Basic Structure:**

```python
class MyClass:
    pass
```

What’s happening:

* `MyClass` is a class.
* But **who creates `MyClass`**?
* The answer: **a metaclass** (usually `type`).

So this:

```python
class MyClass:
    pass
```

Is **really doing** something like this behind the scenes:

```python
MyClass = type("MyClass", (), {})
```

---

## 🛠️ **What Can Metaclasses Do?**

Metaclasses allow you to **control or customize** how classes are created. You can:

* Modify class attributes
* Enforce rules (e.g., naming conventions)
* Automatically register classes
* Inject methods or properties

---

## 🧪 **Example of a Custom Metaclass:**

```python
# A simple metaclass that forces class names to be uppercase
class UppercaseMeta(type):
    def __new__(cls, name, bases, attrs):
        name = name.upper()
        return super().__new__(cls, name, bases, attrs)

class MyClass(metaclass=UppercaseMeta):
    pass

print(MyClass.__name__)  # Output: MYCLASS
```

```python
def singleton(cls):
    instances = {}
    def get_instance(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args, **kwargs)
        return instances[cls]
    return get_instance

class MyClass:
```

```python
        if cls not in cls._instances:

        return cls._instances[cls]

class MyClass(metaclass=SingletonMeta):
    pass
```

## ✅ **What is a Race Condition?**

A **race condition** occurs when **two or more threads (or processes)** access shared data at the same time, and the **final result depends on the timing of their execution**.

It’s called a **"race"** because the threads are **racing** to access and modify the data — and whoever "wins" can lead to **unexpected or incorrect results**.

---

### 🧑‍🎓 **Simple Example:**

Let’s say two threads are adding money to a bank account at the same time:

```python
balance = 0

def deposit():
    global balance
    for _ in range(100000):
        balance += 1

import threading

t1 = threading.Thread(target=deposit)
t2 = threading.Thread(target=deposit)

t1.start()
t2.start()
t1.join()
t2.join()

print(balance)  # Expected: 200000 → But result might be < 200000 😲
```

**Why?**
Because `balance += 1` is not atomic — it's actually 3 steps:

1. Read balance
2. Add 1
3. Write back

If both threads do this at the same time, they **might overwrite each other’s changes**, causing **data corruption**.

---

## 💥 **That’s a race condition.**

It happens when:

* Multiple threads/processes access shared memory
* At least one of them **writes** to it
* There’s no synchronization to control access

---

## 🛡️ **How to Prevent Race Conditions**

### ✅ 1. **Use Locks (`threading.Lock`)** — for multithreading

A lock ensures that **only one thread** can access the critical section at a time.

```python
import threading

balance = 0
lock = threading.Lock()

def deposit():
    global balance
    for _ in range(100000):
        with lock:  # Critical section
            balance += 1

t1 = threading.Thread(target=deposit)
t2 = threading.Thread(target=deposit)

t1.start()
t2.start()
t1.join()
t2.join()

print(balance)  # Now safely gives 200000
```

---

### ✅ 2. **Use `multiprocessing.Lock`** — for multi-processes

If you’re using the `multiprocessing` module, use `multiprocessing.Lock`.

---

### ✅ 3. **Use `threading.RLock`** (reentrant lock) if same thread may acquire lock multiple times.

---

### ✅ 4. **Use Higher-Level Tools**

Python offers thread-safe data structures:

* `queue.Queue`
* `collections.deque` (with `maxlen`)
* `concurrent.futures.ThreadPoolExecutor` (simplifies thread use)

---

### ✅ 5. **Avoid Shared State**

Sometimes the safest design is:

> **Don’t share mutable data between threads/processes at all.**

E.g., use message queues or copy data to avoid races.

---

## 💼 **How to Explain in an Interview**

> **"A race condition happens when multiple threads or processes access and modify shared data at the same time without proper synchronization. The final outcome becomes unpredictable. I typically use locks (`threading.Lock`) to protect critical sections and ensure only one thread can modify shared data at a time. For multi-process applications, I use `multiprocessing.Lock` or move toward lock-free designs with queues and message passing to avoid shared state."**

```python
from collections import defaultdict

def group_anagrams(words):
    anagrams = defaultdict(list)  # Key: sorted word, Value: list of anagrams

    for word in words:
        # Sort the word to use as key — e.g., "tea" → "aet"
        key = ''.join(sorted(word))

        # Add the word to the corresponding anagram group
        anagrams[key].append(word)

    # Return the grouped anagrams as a list of lists
    return list(anagrams.values())
```

## ✅ What is a Context Manager?

A context manager allows you to **set up and tear down resources** automatically using the `with` statement.

Example:

```python
with open("file.txt", "r") as f:
    data = f.read()
```

Here, `open()` is a context manager. It **opens the file**, and after the block is done, it **automatically closes** the file.

## ✅ Two Ways to Write Your Own Context Manager

---

### 🔹 Method 1: Using a Class with `__enter__` and `__exit__`

```python
class MyContext:
    def __enter__(self):
        print("Entering the context...")
        return "Resource"

    def __exit__(self, exc_type, exc_value, traceback):
        print("Exiting the context...")
        if exc_type:
            print(f"Exception caught: {exc_value}")
        return False  # Don't suppress exceptions

# Usage
with MyContext() as res:
    print("Inside the context with:", res)
    # raise ValueError("Oops")  # Try uncommenting to see exception handling
```

### ✅ Output:

```

Inside the context with: Resource
Exiting the context...
```

---

### 🔹 Method 2: Using a Generator + `contextlib.contextmanager`

Simpler and more Pythonic for many use cases:

```python
from contextlib import contextmanager

@contextmanager
def my_context():
    print("Setup: enter")
    yield "Resource"
    print("Cleanup: exit")

# Usage
with my_context() as res:
    print("Using:", res)
```

### ✅ Output:

```

Using: Resource
Cleanup: exit
```

````python
def is_perfect_square(num):
    if num < 0:
        return False
    left, right = 0, num
    while left <= right:
        mid = (left + right) // 2
        print(f"Left: {left}, Right: {right}, Mid: {mid}")
        square = mid * mid
        print(f"Square of {mid} is {square}")
        if square == num:
            return True
        elif square < num:
            left = mid + 1
        else:
            right = mid - 1
    return False

print(is_perfect_square(16))  # True
print(is_perfect_square(14))  # False
````

````python
def remove_duplicates(nums):
    if not nums:
        return 0

    # Initialize the index for the unique elements
    i = 0

    for j in range(1, len(nums)):
        if nums[j] != nums[i]:
            i += 1
            nums[i] = nums[j]  # Move unique element forward

    return i + 1
nums = [1, 1, 2, 3, 3, 4]
length = remove_duplicates(nums)
print("Length:", length)
print("Array after removal:", nums[:length])
````

````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n): # Traverse through all array elements
        for j in range(0, n-i-1): # Last i elements are already in place
            if arr[j] > arr[j+1]: # Traverse the array from 0 to n-i-1
                arr[j], arr[j+1] = arr[j+1], arr[j] # Swap if the element found is greater than the next element
    return arr

#usage
arr = [64, 34, 25, 12, 22, 11, 90]
sorted_arr = bubble_sort(arr)
print("Sorted array is:", sorted_arr)
````
