>
> Subqueries are useful for filtering, checking existence, or comparing values.

```sql
-- Subquery
SELECT name 
FROM Employees 
WHERE department_id IN (SELECT id FROM Departments WHERE name = 'HR');

-- Join
SELECT e.name 
FROM Employees e
JOIN Departments d ON e.department_id = d.id
WHERE d.name = 'HR';
```
