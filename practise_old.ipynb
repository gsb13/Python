````python
def is_balanced(s):
    stack = []
    bracket_map = {')': '(', ']': '[', '}': '{'}
    
    for char in s:
        if char in bracket_map.values():  # Opening brackets
            stack.append(char)
        elif char in bracket_map:         # Closing brackets
            if not stack or stack[-1] != bracket_map[char]:
                return False
            stack.pop()
    
    return not stack 
print(is_balanced("()[]{}"))  # Output: True
print(is_balanced("()[]{")) 
````

````python
from collections.abc import Iterable

def flatten_gen(lst):
    for item in lst:
        if isinstance(item, Iterable) and not isinstance(item, (str, bytes)):
            yield from flatten_gen(item)
        else:
            yield item

# Usage:
print(list(flatten_gen([[1, [2, 3]], 4])))
````

````python
from collections.abc import Iterable

def flatten(obj):
    result = []

    def _flatten(item):
        if item is None:
            return
        if isinstance(item, Iterable) and not isinstance(item, (str, bytes)):
            for sub in item:
                _flatten(sub)
        else:
            result.append(item)

    _flatten(obj)
    return result
````

````python
print(flatten([[1, [2, 3]], 4]))                     # → [1, 2, 3, 4]
print(flatten([1, (2, 3), {4, 5}, "abc"]))           # → [1, 2, 3, 4, 5, 'abc']
print(flatten(["abc", [1, [2, None], (3, 4)]]))      # → ['abc', 1, 2, None, 3, 4]
print(flatten([None, [[], [5, 6]]]))   
````

````python
# write a function to check if a string containing brackets is balanced — 
# meaning every opening bracket has a corresponding closing one in the correct order.

# example: "()", "[]", "{[()]}", "({[]})" are balanced

def is_balanced(string):
    stack = []
    balance = {'(': ')', '[': ']', '{': '}'}
    
    for char in string:
        if char in balance:  # If it's an opening bracket
            stack.append(char)
        elif char in balance.values():  # If it's a closing bracket
            if not stack or balance[stack.pop()] != char:
                return False
    
    return not stack  # If stack is empty, it's balanced

# Example usage:
input_str = '()[]{}'
print(is_balanced(input_str))  # Output: True
````

A **decorator** in programming—particularly in **Python**—is a design pattern that allows you to **add behavior to a function or class without modifying its actual code**. It’s a powerful feature used for **code reuse, separation of concerns, and enhancement of functionality** in a clean, readable way.

---

### 🔧 What Is a Decorator?

In Python, a decorator is a **function that takes another function as an argument**, adds some functionality to it, and returns it.

#### Example:

```python
def decorator_function(original_function):
    def wrapper_function():
        print("Wrapper executed before", original_function.__name__)
        return original_function()
    return wrapper_function

@decorator_function
def say_hello():
    print("Hello!")

say_hello()
```

**Output:**

```
Wrapper executed before say_hello
Hello!
```

---

### 🎯 Significance of Decorators

1. **Separation of concerns**: Keep core logic separate from auxiliary tasks like logging or access control.
2. **Code reusability**: Reuse the same functionality across many functions or methods.
3. **Cleaner syntax**: The `@decorator_name` syntax makes code more readable and maintainable.
4. **DRY principle**: Helps follow "Don't Repeat Yourself" by wrapping repetitive logic.

---

### 🌐 Real-Time Use Cases

#### 1. **Authentication in Web Applications**

Used in frameworks like Flask or Django.

```python
from flask import Flask, request
app = Flask(__name__)

def login_required(f):
    def wrapper(*args, **kwargs):
        if not request.args.get("user"):
            return "Access Denied"
        return f(*args, **kwargs)
    return wrapper

@app.route('/dashboard')
@login_required
def dashboard():
    return "Welcome to Dashboard"
```

#### 2. **Logging**

Used to track function execution without modifying the function itself.

```python
def log_function_call(func):

        return func(*args, **kwargs)
    return wrapper

@log_function_call

```

Decorators can cache results of expensive functions.

```python

```

Used to benchmark or profile code.

````python
# Decorator to cache results
def cache_decorator(func):
    cache_dict = {}

    def wrapper(num):
        if num in cache_dict:
            print(f"Returning cached result for {num}")
            return cache_dict[num]
        else:
            result = func(num)
            cache_dict[num] = result
            print(f"Caching result for {num}")
            return result

    return wrapper

# Use the decorator
@cache_decorator
def square(num):
    return num * num

# Test calls
print(square(10))  # Should calculate and cache
print(square(10))  # Should return cached result
print(square(5))   # Should calculate and cache
print(square(5))   # Should return cached result
print(square(20))  # Should calculate and cache
````

A **generator** in Python is a special type of **iterator** that allows you to iterate over data **without storing the entire dataset in memory**. It’s ideal for working with **large datasets, streaming data, or infinite sequences**.

---

### 🔁 What Is a Generator?

A generator is defined like a normal function but uses the **`yield`** keyword to return values one at a time, **pausing between each**.

#### Example:

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(5):
    print(num)
```

**Output:**

```
1
2
3
4
5
```

The key here is that **`yield` suspends the function’s state**, so it can resume where it left off.

---

### 🎯 Significance of Generators

| Feature              | Benefit                                     |
| -------------------- | ------------------------------------------- |
| **Memory efficient** | Doesn't load all data at once into memory   |
| **Lazy evaluation**  | Computes values on the fly                  |
| **Performance**      | Useful for large or infinite sequences      |
| **Clean syntax**     | Easier than writing custom iterator classes |

---

### 🌐 Real-Time Use Cases of Generators

#### 1. **Reading Large Files Line by Line**

Instead of loading the entire file into memory:

```python
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_large_file("huge_log.txt"):
    print(line)
```

#### 2. **Streaming Data (APIs, IoT, Logs)**

```python
import time

def stream_sensor_data():
    while True:
        yield {"temperature": 25 + (time.time() % 5)}
        time.sleep(1)

for data in stream_sensor_data():
    print(data)
```

#### 3. **Producing Infinite Sequences**

```python
def infinite_counter():
    num = 0
    while True:
        yield num
        num += 1

for i in infinite_counter():
    if i > 10:
        break

```

#### 4. **Database Pagination (Lazy Query Results)**

You can yield rows from a database query without loading all at once.

```python
def fetch_in_batches(cursor, batch_size=100):
    while True:
        records = cursor.fetchmany(batch_size)
        if not records:
            break
        yield from records
```

#### 5. **Pipeline Processing**

Split a large process into stages that pass data using generators.

```python
def tokenize(texts):
    for text in texts:
        yield text.split()

def lowercase(words_stream):
    for words in words_stream:
        yield [w.lower() for w in words]

tokens = lowercase(tokenize(text_list))
for item in tokens:
    print(item)
```

---

### ✅ Summary

| Concept      | Description                                  |
| ------------ | -------------------------------------------- |
| `yield`      | Suspends function execution, returns value   |
| Memory-safe  | Useful for large datasets                    |
| Lazy loading | Only computes what’s needed when it’s needed |
| Reusable     | Easy to chain together with other generators |

````python
# write fibonacci series without using generator
def fibonacci(n):
    a, b = 0, 1
    result = []
    for _ in range(n):
        result.append(a)
        a, b = b, a + b
    return result

fibonacci = fibonacci(10)
for num in fibonacci:
    print(num, end=' ')
    
````

````python
# write fibonacci series using recursion

def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)


for i in range(10):
    print(fibonacci(i), end=' ')
````

### 🧭 What is MRO (Method Resolution Order) in Python?

**MRO (Method Resolution Order)** is the **order in which Python looks for a method or attribute** when it is called on an object, particularly in **multiple inheritance** scenarios.

In simple terms:

> MRO determines **which method is executed first** when multiple classes define the same method name.

---

### 🧠 Why MRO Is Important

* **Avoids ambiguity** in multiple inheritance
* **Ensures consistent method lookup order**
* Prevents **diamond problem** (where multiple parent classes inherit from the same superclass)

---

### 🧪 How to Check MRO

You can check the MRO of a class using:

```python
print(ClassName.__mro__)
# or
ClassName.mro()
```

---

### 🔄 MRO Example with Multiple Inheritance

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

class C(A):
    def show(self):
        print("C")

class D(B, C):
    pass

obj = D()
obj.show()
print(D.mro())
```

#### 🟢 Output:

```
B
[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

### 📌 Explanation:

* Python looks for the method in `D` → then in `B` → then in `C` → then `A` → then `object`.
* It uses the **C3 linearization algorithm** to compute this order consistently.

---

### 🔧 Real-Time Use Case of MRO

#### ✅ 1. **Framework Design (e.g., Django, Flask)**

Frameworks often use **mixins** and multiple base classes to provide reusable functionality:

```python
class LoggingMixin:
    def process(self):
        print("Logging")

class ValidationMixin:
    def process(self):
        print("Validating")

class MyView(LoggingMixin, ValidationMixin):
    def process(self):
        super().process()
        print("Handling request")
```

MRO determines whether `LoggingMixin` or `ValidationMixin`'s `process()` runs first.

---

#### ✅ 2. **Diamond Problem Resolution**

```python
class A:
    def say(self):
        print("A")

class B(A):
    def say(self):
        print("B")

class C(A):
    def say(self):
        print("C")

class D(B, C):
    pass

D().say()
print(D.mro())
```

Even though both `B` and `C` inherit from `A`, Python resolves the method order using MRO to **avoid ambiguity**.

---

### ✅ Summary

| Concept   | Explanation                                                |
| --------- | ---------------------------------------------------------- |
| **MRO**   | Determines order in which classes are searched for methods |
| **Why**   | Helps manage method conflicts in multiple inheritance      |
| **How**   | Python uses **C3 Linearization** to build consistent order |
| **Tools** | Use `ClassName.mro()` or `ClassName.__mro__` to inspect    |

### ⚠️ Exception Handling in Python — Focus on `finally` Block

---

### 🔍 What is `finally`?

The `finally` block in Python is **used to define cleanup actions** that must be executed under all circumstances, **whether an exception occurs or not**.

It’s part of Python's exception handling structure:

```python
try:
    # risky code
except SomeException:
    # handle exception
finally:
    # this code runs no matter what
```

---

### ✅ Key Features of `finally`:

| Feature              | Description                                                            |
| -------------------- | ---------------------------------------------------------------------- |
| Guaranteed execution | Always runs, whether an exception was raised or not                    |
| Resource cleanup     | Commonly used to release resources (e.g., close file or DB connection) |
| Complements `try`    | Can be used with or without `except` block                             |

---

### 📌 Example 1: With Exception

```python
try:
    x = 1 / 0  # raises ZeroDivisionError
except ZeroDivisionError:
    print("Division by zero!")
finally:
    print("This always runs.")
```

**Output:**

```
Division by zero!
This always runs.
```

---

### 📌 Example 2: No Exception

```python
try:
    x = 10 / 2
    print("Result:", x)
finally:
    print("Cleanup done.")
```

**Output:**

```
Result: 5.0
Cleanup done.
```

---

### 📌 Example 3: Closing a File (Real-Time Use Case)

```python
try:
    file = open("data.txt", "r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found.")
finally:
    print("Closing file...")
    file.close()
```

✅ Even if an exception occurs, the file will be **closed** properly to avoid resource leaks.

---

### ⚠️ Real-Time Use Cases

| Scenario | Why Use `finally`?                             |
| -------- | ---------------------------------------------- |
| File I/O | Ensure file is closed after read/write         |
| Database | Ensure DB connection or cursor is closed       |
| Locking  | Release a threading or multiprocessing lock    |
| Logging  | Log exit/cleanup messages regardless of errors |

---

### ✅ Summary

| Block     | Purpose                             |
| --------- | ----------------------------------- |
| `try`     | Code that may throw an exception    |
| `except`  | Code to handle exceptions           |
| `finally` | Code that runs **always** (cleanup) |

---

```python
import sqlite3

try:
    # Try to connect to a (possibly non-existent) database and execute a query

else:
    # Only runs if no exception occurs

    # This will run no matter what

```

| --------- | --------------------------------------------------------------------------- |
| `try`     | Attempts to connect and insert into the database                            |
| `except`  | Catches any database-related errors (like connection issues, bad SQL, etc.) |
| `else`    | Runs only if **no exceptions** were raised — confirms success               |
| `finally` | Ensures the database connection is closed, no matter what                   |

## 🐳 What is Docker?

**Docker** is an **open-source platform** that allows developers to **build, ship, and run applications** in **containers**. A **container** is a lightweight, standalone, and executable software package that includes everything needed to run a piece of software: **code, runtime, system tools, libraries, and settings**.

---

## ⚙️ How to Use Docker?

Here are the basic steps to use Docker:

1. **Install Docker**:

   * Download Docker Desktop from [docker.com](https://www.docker.com/).
   * Install it on your OS (Windows, macOS, Linux).

2. **Create a Dockerfile**:

   * A text file that contains instructions on how to build a Docker image.

   ```Dockerfile
   FROM python:3.10
   COPY . /app
   WORKDIR /app
   RUN pip install -r requirements.txt
   CMD ["python", "app.py"]
   ```

3. **Build an Image**:

   ```bash
   docker build -t my-python-app .
   ```

4. **Run a Container**:

   ```bash
   docker run -p 5000:5000 my-python-app
   ```

5. **Push Image to Docker Hub (Optional)**:

   ```bash
   docker tag my-python-app username/my-python-app
   docker push username/my-python-app
   ```

---

## 🧱 Docker Structure

* **Dockerfile**: Instructions to build an image.
* **Docker Image**: Blueprint for containers.
* **Docker Container**: Running instance of an image.
* **Docker Engine**: The core part that runs containers.
* **Docker Compose**: A tool for defining and running multi-container Docker apps (via `docker-compose.yml`).

---

## ✅ Usefulness of Docker

1. **Portability**:

   * Runs the same on any system (dev, test, production).
2. **Isolation**:

   * Apps run in separate containers with no conflict.
3. **Consistency**:

   * Developers can replicate the same environment.
4. **Scalability**:

   * Easy to scale apps with Docker + orchestration tools (e.g., Kubernetes).
5. **Resource Efficiency**:

   * Uses fewer resources than virtual machines.

## 🌍 Significance of Docker

* **Modern DevOps**: Central tool in CI/CD pipelines.
* **Microservices Architecture**: Each service can run in its own container.
* **Cloud Deployments**: Supported by AWS, Azure, GCP.
* **Standardization**: Industry-wide standard for packaging applications.

## 📦 Class Variables vs. Instance Variables

| Feature       | Class Variable                            | Instance Variable                  |
| ------------- | ----------------------------------------- | ---------------------------------- |
| Belongs to    | The **class itself**                      | Each **object (instance)**         |
| Shared by     | All instances                             | Only the specific instance         |
| Defined in    | Inside the class, **outside** any method  | Usually in the `__init__()` method |
| Accessed with | `ClassName.variable` or `object.variable` | Only via `object.variable`         |

---

## ✅ Example:

```python
class Dog:
    # Class variable
    species = "Canis familiaris"

    def __init__(self, name, age):
        # Instance variables
        self.name = name
        self.age = age

# Create instances
dog1 = Dog("Buddy", 3)
dog2 = Dog("Charlie", 5)

# Access class variable (same for all dogs)
print(dog1.species)  # Canis familiaris
print(dog2.species)  # Canis familiaris

# Access instance variables (different per dog)
print(dog1.name, dog1.age)  # Buddy 3
print(dog2.name, dog2.age)  # Charlie 5
```

---

## 🔄 Changing Variables

### Changing a **class variable**:

```python
Dog.species = "Canis lupus"
print(dog1.species)  # Canis lupus
print(dog2.species)  # Canis lupus
```

### Changing an **instance variable**:

```python
dog1.name = "Max"
print(dog1.name)  # Max
print(dog2.name)  # Charlie
```

## Summary:

| Variable Type         | Where Defined                      | Who Shares It           | How to Access                |
| --------------------- | ---------------------------------- | ----------------------- | ---------------------------- |
| **Class Variable**    | Inside class, outside `__init__()` | Shared by all instances | `ClassName.var` or `obj.var` |
| **Instance Variable** | Inside `__init__()`                | Unique to each instance | `obj.var`                    |

## ✅ Why Use `*args` and `**kwargs` in Decorators?

You use `*args` and `**kwargs` in decorators when:

* You don’t know how many arguments the function being decorated will take.
* You want the decorator to work for **any kind of function**.

---

## 🔧 Basic Decorator with `*args` and `**kwargs`

Here’s a step-by-step example:

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before function call")
        result = func(*args, **kwargs)  # Pass the arguments to the original function
        print("After function call")
        return result
    return wrapper
```

### 🔁 Applying the Decorator:

```python
@my_decorator
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet("Alice", 30)
```

### Output:

```
Before function call
Hello Alice, you are 30 years old.
After function call
```

---

## 🧪 Example with Return Value

```python
def debug(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with args={args} kwargs={kwargs}")
        value = func(*args, **kwargs)
        print(f"{func.__name__} returned {value}")
        return value
    return wrapper

@debug
def add(x, y):
    return x + y

add(5, y=3)
```

---

## 📦 Summary

| Part                    | Purpose                                             |
| ----------------------- | --------------------------------------------------- |
| `*args`                 | Collects all positional arguments                   |
| `**kwargs`              | Collects all keyword arguments                      |
| `func(*args, **kwargs)` | Calls the original function with the same arguments |


## ⏱ 3. **Timing Decorator**

Measures how long a function takes to run.

```python
import time

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Function '{func.__name__}' took {end - start:.4f} seconds")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(2)
    return "Finished!"

slow_function()
```

## 🧠 Visual Memory Model: `is` vs `==`

Imagine you have two objects:

```python
class Person:

        self.name = name

p2 = Person("Alice")
```

Now visualize this in memory:

```
Memory:
+-------------+        +-------------+
|   p1        | -----> |  Person     |
|             |        |  name: Alice|
+-------------+        +-------------+

+-------------+        +-------------+
|   p2        | -----> |  Person     |
|             |        |  name: Alice|
+-------------+        +-------------+
```

Even though the **data** (name = "Alice") is the same, these are two **separate objects** in memory.

---

## 🔍 Code Comparison

```python
print(p1 == p2)  # False, unless you define __eq__
print(p1 is p2)  # False, different objects
```

### Why `==` is False?

By default, Python's `==` compares **identity** for custom objects unless you override `__eq__`. So it's like doing `p1 is p2`.

---

## ✅ Defining `__eq__` to Compare Values

```python
class Person:
    def __init__(self, name):
        self.name = name

    def __eq__(self, other):
        return self.name == other.name

p1 = Person("Alice")
p2 = Person("Alice")

print(p1 == p2)  # True
print(p1 is p2)  # Still False
```

Now:

* `==` compares **values** (`name`)
* `is` still compares **memory location**

---

* Checking identity: `if x is y`
* Checking `None`: `if x is None`

## 📌 Use `==` for:

* Comparing values: `if x == y`

## 1. **List Challenge** — Flatten nested lists

### Approach A: Recursive

```python
def flatten_recursive(lst):
    result = []
    for el in lst:
        if isinstance(el, list):
            result.extend(flatten_recursive(el))
        else:
            result.append(el)
    return result

# Test
print(flatten_recursive([[1,2], [3, [4,5]], 6]))
# → [1, 2, 3, 4, 5, 6]
```

### Approach B: Iterative Stack

```python
def flatten_iterative(lst):
    result, stack = [], lst[::-1]
    while stack:
        el = stack.pop()
        if isinstance(el, list):
            stack.extend(el[::-1])
        else:
            result.append(el)
    return result

# Test
print(flatten_iterative([[1,2], [3, [4,5]], 6]))
# → [1, 2, 3, 4, 5, 6]
```

## 2. **Tuple Challenge** — Partition by predicate

### Approach A: List Comprehension + Convert

```python
def partition_tuple(tup, pred):
    true_group = tuple(x for x in tup if pred(x))
    false_group = tuple(x for x in tup if not pred(x))
    return true_group, false_group

print(partition_tuple((1,2,3,4,5,6), lambda x: x % 2 == 0))
# → ((2, 4, 6), (1, 3, 5))
```

### Approach B: Single-pass Loop

```python
def partition_tuple_one_pass(tup, pred):
    true, false = [], []
    for x in tup:
        (true if pred(x) else false).append(x)
    return tuple(true), tuple(false)

print(partition_tuple_one_pass((1,2,3,4,5,6), lambda x: x % 2))
# → ((2, 4, 6), (1, 3, 5))
```

## 3. **Set Challenge** — Find duplicates in a list

### Approach A: Seen + Duplicates Set

```python
def find_duplicates(lst):
    seen, dups = set(), set()
    for el in lst:
        (dups if el in seen else seen).add(el)
    return dups

print(find_duplicates([1,2,2,3,4,4,4,5]))
# → {2, 4}
```

### Approach B: Counter

```python
from collections import Counter

def find_duplicates_counter(lst):
    return {el for el, cnt in Counter(lst).items() if cnt > 1}

print(find_duplicates_counter([1,2,2,3,4,4,4,5]))
# → {2, 4}
```

## 5. **Mixed Data Structures** — Rotate list by k

### Approach A: Slicing

```python
def rotate_slice(lst, k):
    k %= len(lst)
    return lst[-k:] + lst[:-k]

print(rotate_slice([1,2,3,4,5], 2))
# → [4, 5, 1, 2, 3]
```

### Approach B: Reverse In-place

```python
def reverse(lst, start, end):
    while start < end:
        lst[start], lst[end] = lst[end], lst[start]
        start += 1
        end -= 1

def rotate_inplace(lst, k):
    n = len(lst)
    k %= n
    reverse(lst, 0, n-1)
    reverse(lst, 0, k-1)
    reverse(lst, k, n-1)
    return lst
```

```python
    os.makedirs(backup_dir, exist_ok=True)

    backup_file = os.path.join(backup_dir, f"log_{timestamp}.bak")

    for f in os.listdir(backup_dir):
        path = os.path.join(backup_dir, f)
        if os.path.isfile(path):
            file_time = datetime.fromtimestamp(os.path.getmtime(path))
            if file_time < cutoff:
                os.remove(path)
                print(f"Deleted old backup: {f}")

rotate_backup('app.log', 'log_backups')
```

## 🔹 1. **Constructor (`__init__`) and Destructor (`__del__`)**

### 🔑 What it is:


### ✅ Use Case:

* Managing **resources** like files, network connections, database sessions.

### ⚠️ Drawback:

## 🔹 4. **Class Method vs Static Method**

### 🔑 What it is:

* `@classmethod`: Receives the **class (`cls`)** as its first argument.
* `@staticmethod`: Receives **no class or instance**.

### 🧪 Code:

```python
class Employee:
    raise_percent = 1.10

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    @classmethod
    def set_raise(cls, percent):
        cls.raise_percent = percent

    @staticmethod
    def is_workday(day):
        return day.lower() not in ['saturday', 'sunday']

# Usage
Employee.set_raise(1.20)
print(Employee.raise_percent)  # 1.2
print(Employee.is_workday("Sunday"))  # False
```

### ✅ Use Case:

* `@classmethod`: for **alternative constructors** and modifying class-wide state.
* `@staticmethod`: for **utility methods** that relate to the class but don't need instance or class info.

## 🔹 5. **Multiple Inheritance and MRO (Method Resolution Order)**

### 🔑 What it is:

```python
    def greet(self): print("B")

    def greet(self): print("C")

    pass

d.greet()  # B
print(D.__mro__)
```

### ✅ Use Case:

### ⚠️ Drawback:

* If not carefully designed, can lead to **diamond problems** or confusing behavior.

# 🧠 What is Multithreading?

**Multithreading** is a technique that allows **multiple threads** (smaller units of a process) to **run concurrently**.
Each thread runs independently, but shares memory with other threads of the same process.

---

## ✅ Why Use Multithreading?

* To make programs **more responsive**.
* To perform **I/O-bound** operations faster (e.g., file downloads, web scraping).
* **Not ideal** for CPU-bound tasks due to Python’s GIL (we'll explain this too).

---

## 🔧 Core Concepts

| Concept         | Description                                                                            |
| --------------- | -------------------------------------------------------------------------------------- |
| **Thread**      | A unit of execution within a process.                                                  |
| **Main thread** | The initial thread that runs when your script starts.                                  |
| **GIL**         | Global Interpreter Lock – allows only one thread to execute Python bytecode at a time. |
| **Thread-safe** | Code that behaves correctly when executed by multiple threads.                         |

---

## 🧱 Python’s `threading` Module

## ✅ Example 4: Using ThreadPoolExecutor (Better)

```python
from concurrent.futures import ThreadPoolExecutor
import requests

urls = [
    'https://httpbin.org/image/png',
    'https://httpbin.org/image/jpeg',
    'https://httpbin.org/image/webp',
]

    response = requests.get(url)
    filename = url.split("/")[-1]
    with open(filename, 'wb') as f:
        f.write(response.content)
    return f"{filename} downloaded"

with ThreadPoolExecutor(max_workers=3) as executor:
    results = executor.map(download, urls)

for r in results:
    print(r)
```

🟢 Easier to manage and cleaner than raw threads.

## 🧱 The Problem: GIL (Global Interpreter Lock)

Python’s **GIL** ensures only **one thread** executes Python code at a time (in CPython).
This makes threading **inefficient for CPU-bound tasks** like:

* Large number crunching
* Data processing
* Image processing

✅ Use **`multiprocessing`** instead for CPU-bound tasks.

# 🧠 What is Multiprocessing?

**Multiprocessing** is a technique that allows a program to **run multiple processes** simultaneously — taking advantage of **multiple CPU cores**.

Each process runs **independently**, has its **own memory**, and avoids the **Global Interpreter Lock (GIL)**.

---

## 🔄 Multiprocessing vs. Multithreading

| Feature           | Multithreading                    | Multiprocessing                        |
| ----------------- | --------------------------------- | -------------------------------------- |
| Memory            | Shared between threads            | Separate memory per process            |
| Speed (CPU-bound) | Slower due to GIL                 | Faster – uses multiple cores           |
| Best for          | I/O-bound tasks (e.g., file, web) | CPU-bound tasks (e.g., data crunching) |
| Crashes           | One thread crash = app lives      | One process crash = isolated           |

---

## ✅ When to Use Multiprocessing

* Heavy **data processing**
* **Parallel tasks** that don’t need to share memory
* Tasks that are CPU-intensive (e.g., image processing, ML, simulations)

---

## 🔧 Core Module: `multiprocessing`

Main tools:

* `Process` – run a function in a new process
* `Pool` – manage a pool of worker processes
* `Queue`, `Pipe` – for inter-process communication (IPC)
* `Lock` – to synchronize access to shared resources

---

## ✅ Example 1: Basic Process

```python
from multiprocessing import Process
import time

def task(name):
    print(f"Process {name} starting...")
    time.sleep(2)
    print(f"Process {name} done.")

if __name__ == '__main__':
    p1 = Process(target=task, args=("A",))
    p2 = Process(target=task, args=("B",))

    p1.start()
    p2.start()

    p1.join()
    p2.join()
    print("Both processes complete.")
```

---

## ✅ Example 2: Using `Pool` for Parallel Map

```python
from multiprocessing import Pool
import time

def square(n):
    time.sleep(1)
    return n * n

if __name__ == '__main__':
```

* Tasks are divided and run **in parallel**.

## ✅ Shared State (with `Value` and `Array`)

```python
from multiprocessing import Process, Value
import time

def increment(counter):
    for _ in range(100000):
        counter.value += 1

if __name__ == '__main__':
    count = Value('i', 0)
    p1 = Process(target=increment, args=(count,))
    p2 = Process(target=increment, args=(count,))

    p1.start()
    p2.start()
    p1.join()
    p2.join()
```

## ✅ Using a Lock to Avoid Race Conditions

```python
from multiprocessing import Process, Value, Lock

def safe_increment(counter, lock):
    for _ in range(100000):
        with lock:
            counter.value += 1

if __name__ == '__main__':
    counter = Value('i', 0)
    lock = Lock()

    p1 = Process(target=safe_increment, args=(counter, lock))
    p2 = Process(target=safe_increment, args=(counter, lock))

    p1.start()
    p2.start()
    p1.join()
    p2.join()

    print("Safe count:", counter.value)
```

# 🔄 What is `asyncio`?

`asyncio` is a built-in Python module that provides the infrastructure to write **concurrent code using the async/await syntax**.

---

## ✅ Key Difference from Threading and Multiprocessing

| Feature     | Threading       | Multiprocessing           | `asyncio`                               |
| ----------- | --------------- | ------------------------- | --------------------------------------- |
| Best for    | I/O-bound tasks | CPU-bound tasks           | I/O-bound tasks                         |
| Concurrency | OS threads      | OS processes              | Coroutine (lightweight tasks)           |
| Memory      | Shared          | Separate                  | Shared                                  |
| GIL Impact  | Blocked by GIL  | Avoids GIL with processes | Avoids GIL via cooperative multitasking |
| Overhead    | Moderate        | High                      | Very Low                                |

---

## 🔧 When to Use `asyncio`

* High-performance **web servers or APIs**
* **Web scraping** (non-blocking HTTP)
* **Chat apps**, **sockets**, **real-time systems**
* File I/O (with third-party libs like `aiofiles`)
* Avoids using multiple threads/processes for I/O

---

## 🔑 Key Concepts

| Concept         | Description                          |
| --------------- | ------------------------------------ |
| `async def`     | Defines an asynchronous coroutine    |
| `await`         | Waits for the result of a coroutine  |
| `asyncio.run()` | Starts the main coroutine            |
| `create_task()` | Run multiple coroutines concurrently |
| `sleep()`       | Non-blocking sleep                   |

---

## ✅ Basic Example: Async Sleep

```python
import asyncio

async def greet(name):
    print(f"Hello {name}")
    await asyncio.sleep(1)
    print(f"Goodbye {name}")

async def main():
    await asyncio.gather(
        greet("Alice"),
        greet("Bob"),
    )

asyncio.run(main())
```

⏱ Total time ≈ 1 second, **not 2**, because the coroutines run concurrently.

## ✅ Real-World Example: Web Requests

```python
import asyncio
import aiohttp  # pip install aiohttp

urls = [
    "https://httpbin.org/delay/1",
    "https://httpbin.org/delay/2",
    "https://httpbin.org/delay/3"
]

async def fetch(session, url):
    async with session.get(url) as response:
        print(f"Fetched {url}")
        return await response.text()

async def main():
    async with aiohttp.ClientSession() as session:
        tasks = [fetch(session, url) for url in urls]
        responses = await asyncio.gather(*tasks)
        print(f"Got {len(responses)} responses")

asyncio.run(main())
```

✅ **Faster than synchronous requests**, perfect for **scraping APIs**.

## ✅ Async Queue for Background Processing

```python
import asyncio

async def producer(queue):
    for i in range(5):
        print(f"Producing {i}")
        await queue.put(i)
        await asyncio.sleep(0.5)

async def consumer(queue):
    while True:
        item = await queue.get()
        print(f"Consumed {item}")
        queue.task_done()

async def main():
    queue = asyncio.Queue()
    producer_task = asyncio.create_task(producer(queue))
    consumer_task = asyncio.create_task(consumer(queue))

    await producer_task
    await queue.join()
    consumer_task.cancel()

asyncio.run(main())
```

✅ Great for **background job queues**, **pipeline systems**, **stream processing**.

# 🧵 `threading`

## ✅ What is it?

* Allows **concurrent execution** of tasks using multiple **threads** within a single process.
* Ideal for **I/O-bound** tasks (e.g., reading files, making API calls).

## ⚙️ How it works

* Multiple threads share the **same memory** space.
* Python threads are limited by the **Global Interpreter Lock (GIL)** – only one thread runs Python bytecode at a time.

## 🎯 When to use

* Tasks that **wait** a lot (e.g., network, disk I/O)
* When you don’t need heavy CPU usage

## 🔧 Real-World Use Cases

* Web crawlers
* File uploaders
* GUI background tasks
* Logging systems

## ⚠️ Limitations

* **Not suitable for CPU-bound tasks**
* Can cause **race conditions** — needs `Lock`, `Semaphore`, etc.

---

# 🧠 `multiprocessing`

## ✅ What is it?

* Runs **multiple processes** in parallel, each with **its own Python interpreter**.
* Great for **CPU-bound** tasks like image or data processing.

## ⚙️ How it works

* Each process has its **own memory space**.
* Can **truly parallelize** tasks across CPU cores.
* Avoids the GIL altogether.

## 🎯 When to use

* CPU-heavy operations
* Data crunching or transformations
* Running tasks in complete isolation

## 🔧 Real-World Use Cases

* Image processing (OpenCV, PIL)
* Data science (pandas, numpy calculations)
* ML model training
* Batch file conversions

## ⚠️ Limitations

* **High memory** consumption (copies data)
* Inter-process communication (IPC) is **harder/slower**
* More startup/shutdown overhead

---

# 🌀 `asyncio`

## ✅ What is it?

* A Python module for **asynchronous, cooperative multitasking**.
* Best for **I/O-bound** and **high concurrency** tasks (like thousands of HTTP requests).

## ⚙️ How it works

* Runs on a **single thread**, uses an **event loop**.
* Tasks yield control using `await` so others can run.

## 🎯 When to use

* High-concurrency apps (e.g., websockets, REST APIs)
* Thousands of small I/O tasks (network, file)

## 🔧 Real-World Use Cases

* Web scraping (aiohttp)
* Async file operations (aiofiles)
* Chat servers
* Streaming and notification systems

## ⚠️ Limitations

* **Not ideal for CPU-bound tasks**
* Doesn’t use multiple CPU cores
* Doesn’t preempt tasks — your coroutines must behave

# ⚙️ 1. What is an Iterator?

An **iterator** is an object that lets you **iterate over a sequence** one item at a time.

### ✅ Iterator Protocol

An object is an **iterator** if it:

* Implements `__iter__()` → returns itself
* Implements `__next__()` → returns next value or raises `StopIteration`

```python
class Counter:
    def __init__(self, limit):
        self.current = 0
        self.limit = limit

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.limit:
            val = self.current
            self.current += 1
            return val
        else:
            raise StopIteration

counter = Counter(3)
for num in counter:
    print(num)  # 0, 1, 2
```

# 🔄 2. `iter()` and `next()`

* `iter(obj)` calls `obj.__iter__()`
* `next(obj)` calls `obj.__next__()`

```python
nums = [10, 20, 30]
it = iter(nums)
print(next(it))  # 10
print(next(it))  # 20
print(next(it))  # 30
# next(it) would raise StopIteration
```

# ⚡ 3. Generator Functions: `yield`

A **generator** is a simpler way to create iterators using `yield`.

### ✅ Example:

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(3):
    print(num)  # 1, 2, 3
```

* `yield` **pauses** function state.
* Next call resumes **from where it left off**.

---

# 🔁 4. `yield from`: Delegate to sub-generator

Useful when breaking a generator into **smaller parts**.

```python
def sub():
    yield 1
    yield 2

def main():
    yield 0
    yield from sub()
    yield 3

for i in main():
    print(i)  # 0 1 2 3
```

# 🧠 5. Generator Expressions (One-liner)

Just like list comprehensions, but lazy:

```python
squares = (x*x for x in range(5))
print(next(squares))  # 0
print(next(squares))  # 1
```

✅ **Saves memory** compared to `[x*x for x in range(1_000_000)]`

````python
'''
Withot using any libraries, we can implement the normalization manually.
This includes converting characters to lowercase, capitalizing the first letter of each word,
and ensuring that words are split and joined correctly. 
'''

# Convert a character to lowercase (ASCII-based)
def to_lower(char):
    if 'A' <= char <= 'Z':
        return chr(ord(char) + 32)
    return char

# Convert a character to uppercase (ASCII-based)
def to_upper(char):
    if 'a' <= char <= 'z':
        return chr(ord(char) - 32)
    return char

# Convert entire word to lowercase
def word_to_lower(word):
    result = ''
    for ch in word:
        result += to_lower(ch)
    return result

# Capitalize word: First letter uppercase, rest lowercase
def capitalize_word(word):
    if word == '':
        return ''
    first = to_upper(word[0])
    rest = ''
    for ch in word[1:]:
        rest += to_lower(ch)
    return first + rest

# Split sentence into words (manual split on space)
def split_words(text):
    words = []
    current = ''
    for ch in text:
        if ch == ' ':
            if current != '':
                words.append(current)
                current = ''
        else:
            current += ch
    if current != '':
        words.append(current)
    return words

# Join words with single space
def join_words(words):
    result = ''
    for i in range(len(words)):
        result += words[i]
        if i != len(words) - 1:
            result += ' '
    return result

# Final normalization function
def normalize_command(text):
    words = split_words(text)
    result_words = []
    for word in words:
        word_lower = word_to_lower(word)
        word_cap = capitalize_word(word_lower)
        result_words.append(word_cap)
    return join_words(result_words)

# Input
inputs = [
    'Navigate to starbucks',
    'navigate to sTarbUcks',
    'NAVIGATE to STARBUCKS'
]

# Output (with deduplication)
output = []
seen = []

for phrase in inputs:
    normalized = normalize_command(phrase)
    if normalized not in seen:
        seen.append(normalized)
        output.append(normalized)

# Print result
for line in output:
    print(line)
````

### 🔍 Explanation:

* `D` inherits from `A`, `B`, and `C`.

* Each class uses `super().show()` to call the next method in the MRO chain.

* Use `super()` consistently to make cooperative multiple inheritance work correctly.
* Always check MRO with `ClassName.__mro__` if you're unsure of method resolution.

````python
def add_show(cls):
    # Define the cooperative show() method
    def show(self):
        print(f"This is {cls.__name__} class")
        # Try to call super().show() if it exists
        try:
            super(cls, self).show()
        except AttributeError:
            pass  # Base class with no show()

    # Inject method into class
    cls.show = show
    return cls


@add_show
class A:
    def __init__(self):
        print("Init A")


@add_show
class B(A):
    def __init__(self):
        print("Init B")
        super().__init__()


@add_show
class C(A):
    def __init__(self):
        print("Init C")
        super().__init__()


@add_show
class D(B, C):
    def __init__(self):
        print("Init D")
        super().__init__()

obj = D()
obj.show()
````

---

## The Problem We’re Solving

You have multiple classes in a hierarchy, and you want each class to have a `show()` method that:

* Prints something like `"This is <ClassName> class"`.
* Calls the parent class’s `show()` method automatically (if it exists), so all messages from the class hierarchy get printed in order.

You want to **avoid writing the `show()` method in every class manually**, so we use a **decorator** to add it automatically.

---

## The Decorator Code

```python
def add_show(cls):
    def show(self):
        print(f"This is {cls.__name__} class")  # Print the current class name
        
        # Call the next class's show() in the MRO chain if it exists
        try:
            super(cls, self).show()
        except AttributeError:
            pass  # No parent show() method to call (base case)

    cls.show = show  # Inject this show method into the class
    return cls       # Return the modified class
```

---

### What is going on here?

1. **`def add_show(cls):`**

   This is the decorator function — it takes a class `cls` as input.

2. **`def show(self):`**

   Defines a new method `show()` to add to the class dynamically.

3. **`print(f"This is {cls.__name__} class")`**

   Prints the message using the class name (`cls.__name__`).

4. **`super(cls, self).show()`**

   * This is a key line.
   * It calls the **next class’s** `show()` method in the MRO *after* `cls`.
   * Normally, `super()` is used inside methods, but here we specify `cls` explicitly to tell Python:
     "Find the next class after `cls` in the MRO for this instance `self`."
   * This enables chaining all `show()` methods in the hierarchy, respecting MRO order.

5. **`except AttributeError: pass`**

   If no `show()` method exists in the next class (usually the last class, like `object`), just ignore the error and stop calling further.

6. **`cls.show = show`**

   Dynamically attaches this new `show` method to the class `cls`.

7. **`return cls`**

   Returns the modified class back, so it replaces the original class definition.

---

## How You Use It

```python
@add_show
class A:
    def __init__(self):
        print("Init A")
```

* When Python reads this, it calls `add_show(A)`.
* The decorator adds the `show()` method to `A` automatically.
* You do the same for `B`, `C`, and `D`.

---

## What Happens When You Call `obj.show()` on `D`?

1. The `show()` method added by the decorator for `D` runs:

   * Prints `"This is D class"`
   * Calls `super(D, obj).show()` → calls `B`’s `show()` (next in MRO).

2. `B`’s decorator-added `show()` runs:

   * Prints `"This is B class"`
   * Calls `super(B, obj).show()` → calls `C`’s `show()`.

3. `C`’s decorator-added `show()` runs:

   * Prints `"This is C class"`
   * Calls `super(C, obj).show()` → calls `A`’s `show()`.

4. `A`’s decorator-added `show()` runs:

   * Prints `"This is A class"`
   * Calls `super(A, obj).show()` → no more `show()`; catches `AttributeError` and stops.

## ✅ Example 1: Get Public Repos for a User

```python
import requests

def get_repos_for_user(username):
    url = f"https://api.github.com/users/{username}/repos"
    response = requests.get(url)
    
    if response.status_code != 200:
        print("Error:", response.status_code)
        return

    repos = response.json()

    for repo in repos:
        name = repo['name']
        owner = repo['owner']['login']
        full_name = repo['full_name']
        print(f"Name: {name}, Owner: {owner}, Full Name: {full_name}")
```

### ✅ 2. Python Code to Fetch Private Repos

```python
import requests

def get_private_repos(token):
    url = "https://api.github.com/user/repos"
    headers = {
        "Authorization": f"token {token}"
    }

    params = {
        "visibility": "private"  # Optional: filter only private repos
    }

    response = requests.get(url, headers=headers, params=params)

    if response.status_code != 200:
        print("Error:", response.status_code, response.text)
        return

    repos = response.json()

    for repo in repos:
        name = repo['name']
        owner = repo['owner']['login']
        full_name = repo['full_name']
        print(f"Name: {name}, Owner: {owner}, Full Name: {full_name}")

# 🔐 Replace with your actual GitHub token
token = "ghp_xxx...your_token_here..."
get_private_repos(token)
```
