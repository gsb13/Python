1. Bar Chart (Categorical or Ordinal → Numerical)

Use when you want to show the comparison between categories.

Works well for both discrete numerical data and counts.

✅ Use When:

You have categories and want to compare their values.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# 2. Pie Chart (Categorical → Proportion)
# Shows proportions of categories as slices of a circle.
# ✅ Use When:
# You want to show relative parts of a whole.

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]

plt.pie(sizes, labels=labels, autopct='%1.2f%%')
plt.title('Pie Chart Example')
plt.show()
```

```python
# 3. Histogram (Numerical Continuous → Frequency)
# Good for visualizing the distribution of continuous data.
# ✅ Use When:
# You want to understand the spread or frequency of values.

import numpy as np

data = np.random.randn(1000)  # Normally distributed data

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Line Chart (Ordinal or Time → Continuous)
# Best for time-series or ordered data.
# ✅ Use When:
# You want to show trends over time or ordered categories.

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
values = [20, 22, 19, 24, 27]

plt.plot(days, values, marker='o')
plt.title('Line Chart Example')
plt.xlabel('Day')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# 5. Scatter Plot (Numerical vs Numerical)

# Used to see relationships/correlations between two continuous variables.

# ✅ Use When:

# You want to visualize the correlation or distribution between two numeric variables.

x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)

plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# . Box Plot (Categorical → Distribution)

# Summarizes the distribution of a dataset, showing median, quartiles, and outliers.

# ✅ Use When:

# You want to compare distributions across different categories.

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Box Plot Example')
plt.ylabel('Value')
plt.show()
```

```python
# 7. Heatmap (Categorical/Categorical → Numerical)

# Displays data in a matrix form with color gradients.

# ✅ Use When:

# You want to show interactions or correlation matrices.

import seaborn as sns
import pandas as pd

# Generate random correlation matrix
data = np.random.rand(5, 5)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

sns.heatmap(df, annot=True, cmap='coolwarm')
plt.title('Heatmap Example')
plt.show()
```

📘 Summary Table
Chart Type	X-Axis Data Type	Y-Axis Data Type	Best For
Bar Chart	Categorical/Ordinal	Numerical	Comparing categories
Pie Chart	Categorical	Proportions	Showing part-of-whole
Histogram	Numerical (bins)	Frequency	Distribution of data
Line Chart	Ordinal/Time	Numerical	Trends over time
Scatter Plot	Numerical	Numerical	Relationships between variables
Box Plot	Categorical	Numerical	Distribution + outliers
Heatmap	Categorical/Categorical	Numerical	Matrix-like data (e.g., correlations)
