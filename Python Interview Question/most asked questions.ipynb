# 2. Check Palindrome 

```python
# Check Palindrome 
s = "madam"
rev = ""

for ch in s:
    rev = ch + rev

print(s == rev)
```

```python
def check_palindrome(input_string):
    reversed_str = ""
    for char in input_string:
        reversed_str = char + reversed_str
    return input_string == reversed_str

check_palindrome("racecar")
```

```python
# 3. Count Character Frequency

def count_char_freq(input_string):
    freq_dict = {}
    for char in input_string:
        if char in freq_dict:
            freq_dict[char] += 1
        else:
            freq_dict[char] = 1
    return freq_dict
s = "hello world"
count_char_freq(s)


```

# ðŸš€ 4. First Non-Repeating Character

```python
#First Non-Repeating Character without inbuilt function
def first_non_repeating_char(input_string):
    char_count = {}
    for char in input_string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    for char in input_string:
        if char_count[char] == 1:
            return char
    return None

first_non_repeating_char("swiss")
```

```python
# with inbuilt function
from collections import Counter

s = "swiss"
freq = Counter(s)

print(next(ch for ch in s if freq[ch] == 1))
```

# ðŸš€ 5. Fibonacci Series

```python
def fibonacci_series(n):
    fib_sequence = []
    a, b = 0, 1
    for _ in range(n):
        fib_sequence.append(a)
        a, b = b, a + b
    return fib_sequence

fibonacci_series(10)
```

# ðŸš€ 6. Check Prime Number

```python
def check_prime_number(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1): #int(num**0.5) + 1 -  
        if num % i == 0:
            return False
    return True 
check_prime_number(29)
print('the given number is prime')
```

```python
# 7. Second Largest Number in List
def second_largest_number(num_list):
    first = second = float('-inf')
    for num in num_list:
        if num > first:
            second = first
            first = num
        elif first > num > second:
            second = num
    return second if second != float('-inf') else None

second_largest_number([10, 20, 4, 54, 99, 99])
```

# 8. Remove Duplicates from List


```python
def remove_duplicates(input_list):
    return list(set(input_list))
remove_duplicates([1, 2, 2, 3, 4, 4, 5])
```

```python
# with programming logic
def remove_duplicates(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

remove_duplicates([1, 2, 2, 3, 4, 4, 5])
```

# 9. Reverse a List

```python
#  with slicing
def reverse_list(input_list):
    return input_list[::-1]

reverse_list([1, 2, 3, 4, 5])
```

```python
# programming logic
def reverse_list(input_list):
    reversed_list = []
    for item in input_list:
        reversed_list.insert(0, item)
    return reversed_list
reverse_list([1, 2, 3, 4, 5])
```

```python
# other method 
def reverse_list_new(input_list):
    reversed_list = []
    for i in range(len(input_list)-1, -1, -1):
        reversed_list.append(input_list[i])
    return reversed_list

reverse_list_new([1, 2, 3, 4, 5])
```

# 10. Run-Length Encoding


```python
def run_length_encoding(text):
    result = ""
    count = 1

    for i in range(len(text)):
        if i < len(text) - 1 and text[i] == text[i + 1]: 
            # check next character is same or not 
            count += 1
        else:
            result += str(count) + text[i]
            count = 1
    return result
text = "aaabbca"
result = run_length_encoding(text)
print(result)
```

# ðŸš€ 11. Find Missing Number in Sequence

```python
#Find Missing Number in Sequence

def find_missing_number(num_list, n):
    """ logic to find missing number in a sequence is as follows 
     1. Calculate the expected sum of first n natural numbers using the formula n*(n+1)/2
     2. Calculate the actual sum of the numbers present in the list
     3. The missing number is the difference between the expected sum and the actual sum
    """
    expected_sum = n * (n + 1) // 2 # formula to calculate sum of first n natural numbers
    actual_sum = sum(num_list) 
    return expected_sum - actual_sum

find_missing_number([1, 2, 4, 5, 6], 6)
```

# ðŸš€ 12. Check Anagram

```python
# check anagram
def are_anagrams(str1, str2):
    return sorted(str1) == sorted(str2)

are_anagrams("listen", "silent")
```

```python
# with programming logic
def are_anagrams(str1, str2):
    """logic to check anagram is as follows
    1. If the lengths of the two strings are different, they cannot be anagrams.
    2. Create a dictionary to count the frequency of each character in the first string.
    3. Iterate through the second string, decrementing the character counts in the dictionary.
    4. If any character in the second string is not found in the dictionary or its count goes below zero, the strings are not anagrams.
    5. If all character counts return to zero, the strings are anagrams."""
    if len(str1) != len(str2):
        return False

    char_count = {}

    for char in str1:
        char_count[char] = char_count.get(char, 0) + 1

    for char in str2:
        if char not in char_count or char_count[char] == 0:
            return False
        char_count[char] -= 1

    return True

are_anagrams("listen", "silent")
```
