# üöÄ 11. Find Missing Number in Sequence

````python
#Find Missing Number in Sequence

def find_missing_number(num_list, n):
    """ logic to find missing number in a sequence is as follows 
     1. Calculate the expected sum of first n natural numbers using the formula n*(n+1)/2
     2. Calculate the actual sum of the numbers present in the list
     3. The missing number is the difference between the expected sum and the actual sum
    """
    expected_sum = n * (n + 1) // 2 # formula to calculate sum of first n natural numbers
    actual_sum = sum(num_list) 
    return expected_sum - actual_sum

find_missing_number([1, 2, 4, 5, 6], 6)
````

# üöÄ 12. Check Anagram

````python
# check anagram
def are_anagrams(str1, str2):
    return sorted(str1) == sorted(str2)

are_anagrams("listen", "silent")
````

````python
# with programming logic
def are_anagrams(str1, str2):
    """logic to check anagram is as follows
    1. If the lengths of the two strings are different, they cannot be anagrams.
    2. Create a dictionary to count the frequency of each character in the first string.
    3. Iterate through the second string, decrementing the character counts in the dictionary.
    4. If any character in the second string is not found in the dictionary or its count goes below zero, the strings are not anagrams.
    5. If all character counts return to zero, the strings are anagrams."""
    if len(str1) != len(str2):
        return False

    char_count = {}

    for char in str1:
        char_count[char] = char_count.get(char, 0) + 1

    for char in str2:
        if char not in char_count or char_count[char] == 0:
            return False
        char_count[char] -= 1

    return True

are_anagrams("listen", "silent")
````

# ‚úÖ **1. Singleton Pattern ‚Äî Only ONE instance allowed**

### **Meaning**

A class is Singleton when only **one object/instance** can ever be created from it.

### **Why needed?**

Some components must exist only once in the system:

* One database connection for whole app
* One logging manager
* One configuration reader
* One cache object

Creating multiple objects may cause memory problems or inconsistent states.

### **How it works?**

Python allows you to override object creation using `__new__()`.
We store the instance in a class variable.

### **Code (with explanation)**

```python
class Singleton:
    """
    Singleton pattern ensures only one instance exists.
    """
    _instance = None  # stores the single object

    def __new__(cls):
        if cls._instance is None:      # if no instance is created
            cls._instance = super().__new__(cls)
        return cls._instance           # return same instance every time
```

### **Explanation (step-by-step)**

1. Python calls `__new__()` *before* `__init__()`.
2. First time ‚Üí `_instance` is `None` ‚Üí new object created.
3. Second time ‚Üí `_instance` already has object ‚Üí returns same object.

So both `a` and `b` point to the same memory location.

# ‚úÖ **2. Factory Pattern ‚Äî Class that creates objects**

### **Meaning**

Factory Pattern helps create objects **without exposing logic** to the user.

Instead of doing:

```python

```

You do:


### **Why needed?**

* Removes complex `if/else` from code
* Adds new object types easily
* Used in frameworks (Django ORM, SQLAlchemy, Selenium drivers)

### **Real-world example**

You book a ride in Uber:
You select *Auto*, *Bike*, or *Car*.
You don‚Äôt create the vehicle object.
The system‚Äôs *Factory* does it.

### **Code**

```python
class ShapeFactory:

    def get_shape(shape_type):

            return Circle()
        elif shape_type == "square":
            return Square()
```

### **Flow**

1. User requests object by name
2. Factory uses logic to create right class object
3. User gets object without knowing creation details

# üî• PART 1 ‚Äî MULTITHREADING & MULTIPROCESSING (DETAILED)

---

# ‚úÖ **1. Multithreading (I/O-bound tasks)**

### **Meaning**

Running *multiple threads* inside the same process.

Threads share:

* Memory
* Variables
* File handles

### **Best used for (I/O-bound tasks)**

‚úî File read/write
‚úî Database queries
‚úî API calls
‚úî Network operations

### **Why threads don‚Äôt speed up CPU tasks?**

Because of CPython‚Äôs **GIL (Global Interpreter Lock)**
‚Üí Only *one* thread executes Python bytecode at a time.

### **Example**

```python
import threading
import time

def task():
    print("Task started")
    time.sleep(2)
    print("Task completed")

t1 = threading.Thread(target=task)
t1.start()
t1.join()
```

### **Flow**

* Thread starts
* Doesn‚Äôt block main thread
* `join()` waits until thread finishes

---

# ‚úÖ **2. Multiprocessing (CPU-bound tasks)**

### **Meaning**

Runs *multiple processes*, each with separate memory and its own Python interpreter ‚Üí bypasses GIL.

### **Best used for (CPU-bound tasks)**

‚úî Machine Learning calculations
‚úî Image/video processing
‚úî Data transformations
‚úî Math-heavy loops

### **Example**

```python
from multiprocessing import Process

def compute():
    print("Process running")

p = Process(target=compute)
p.start()
p.join()
```

### **Why multiprocessing is faster for CPU tasks?**

Multiple CPU cores ‚Üí each process executes independently.

# ‚ùì **What is an Iterator?**

An object that you can loop over using `next()`.

### **Python iterator protocol**

```python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.value = 0

    def __iter__(self):
        return self

    def next(self):            # For Python2
        pass

    def __next__(self):        # For Python3
        if self.value < self.limit:
            self.value += 1
            return self.value
        raise StopIteration
```

### **Where used?**

* Generators
* File objects
* Database cursors
* Data streams

# üîπ SOLID PRINCIPLES ‚Äî Simple & Technical Explanation

---

## **S ‚Äî Single Responsibility Principle (SRP)**

**Technical meaning:**
A class should **do only one thing**. Don‚Äôt mix multiple jobs in the same class.

**Analogy for a school boy:**
Imagine a **teacher**. The teacher teaches **math** only.
If the teacher also cooks lunch, fixes the classroom, and drives the bus, it becomes messy.

**Python Example**

```python
# ‚ùå Bad ‚Äî doing too many things
class Student:
    def __init__(self, name):
        self.name = name

    def attend_class(self):
        print(f"{self.name} is attending class")

    def submit_homework(self):
        print(f"{self.name} submits homework")

    def pay_fees(self):
        print(f"{self.name} pays fees")  # Too many responsibilities
```

```python
# ‚úÖ Good ‚Äî single responsibility
class Attendance:
    def mark(self, student):
        print(f"{student} attended class")

class Homework:
    def submit(self, student):
        print(f"{student} submitted homework")

class Payment:
    def pay(self, student):
        print(f"{student} paid fees")
```

**Key idea:** Each class has **one responsibility ‚Üí easier to maintain**.

---

## **O ‚Äî Open/Closed Principle (OCP)**

**Technical meaning:**
Classes should be **open for extension, closed for modification**.
You **don‚Äôt change existing code**, just add new features.

**Analogy:**
A **video game console**. You can add new games (extend), but you don‚Äôt change the console‚Äôs hardware.

**Python Example**

```python
# ‚ùå Bad ‚Äî modifying code for new shapes
class AreaCalculator:
    def calculate_area(self, shape):
        if shape.type == "circle":
            return 3.14 * shape.radius**2
        elif shape.type == "square":
            return shape.side**2
```

```python
# ‚úÖ Good ‚Äî extend by adding new class
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius**2

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side**2

# Now you can add Triangle without modifying previous classes
```

**Key idea:** Add new things without breaking old code.

---

## **L ‚Äî Liskov Substitution Principle (LSP)**

**Technical meaning:**
If **B is a child of A**, B should behave like A.
You should be able to replace A with B without breaking the code.

**Analogy:**
A **car** is a vehicle. If you replace the car with a **bike**, it should still move on the road like a vehicle.
If your ‚Äúbike‚Äù can‚Äôt move or breaks all rules of a vehicle ‚Üí it violates LSP.

**Python Example**

```python
class Bird:
    def fly(self):
        print("Bird can fly")

class Eagle(Bird):
    pass

class Ostrich(Bird):
    def fly(self):
        raise Exception("Ostrich can't fly")  # ‚ùå violates LSP
```

**Better design:**

```python
class Bird:
    def eat(self):
        print("Bird is eating")

class FlyingBird(Bird):
    def fly(self):
        print("Flying bird can fly")

class Eagle(FlyingBird):
    pass

class Ostrich(Bird):
    pass  # no fly method, still works fine
```

**Key idea:** Child class should **behave like parent**.

---

## **I ‚Äî Interface Segregation Principle (ISP)**

**Technical meaning:**
Don‚Äôt force a class to implement methods it **doesn‚Äôt need**.

**Analogy:**
Imagine a **smartphone** interface. It has:

* Call
* Camera
* GPS
* Fax

A student phone doesn‚Äôt need Fax. Don‚Äôt force it to implement Fax.

**Python Example**

```python
# ‚ùå Bad ‚Äî class forced to implement unnecessary methods
class Machine:
    def print(self): pass
    def scan(self): pass
    def fax(self): pass

class Printer(Machine):
    def print(self): pass
    def scan(self): pass
    def fax(self):
        raise Exception("Printer can't fax")  # ‚ùå problem
```

```python
# ‚úÖ Good ‚Äî segregate interfaces
class PrinterInterface:
    def print(self): pass

class ScannerInterface:
    def scan(self): pass

class FaxInterface:
    def fax(self): pass

class Printer(PrinterInterface):
    def print(self):
        print("Printing...")
```

**Key idea:** Keep interfaces **small & relevant**.

---

## **D ‚Äî Dependency Inversion Principle (DIP)**

**Technical meaning:**
High-level modules should **not depend on low-level modules**.
Both should depend on **abstraction**.

**Analogy:**
A **robot** that moves:

* High-level: ‚ÄúMove forward‚Äù
* Low-level: motor details
  Robot should use motor **interface**, not direct motor code.

**Python Example**

```python
# ‚ùå Bad ‚Äî High-level depends on low-level
class MySQLDatabase:
    def connect(self):
        print("Connected to MySQL")

class UserService:
    def __init__(self):
        self.db = MySQLDatabase()  # ‚ùå tight coupling
```

```python
# ‚úÖ Good ‚Äî depends on abstraction
class Database:
    def connect(self): pass

class MySQLDatabase(Database):
    def connect(self):
        print("Connected to MySQL")

class UserService:
    def __init__(self, db: Database):
        self.db = db  # ‚úÖ loose coupling
```

**Key idea:** High-level code depends on **interface**, not concrete implementation.
